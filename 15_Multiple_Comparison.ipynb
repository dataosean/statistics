### 1. 다중 비교 문제의 개요

1) 다중 비교의 정의

    - 다중 비교(Multiple Comparison)는 여러 개의 가설을 동시에 검정하는 과정을 말한다.
    
    - 단일 가설 검정에서 유의수준(α)을 설정해 1종 오류(잘못된 기각)를 제어할 수 있지만,
    
    - 다중 비교에서는 여러 가설을 동시에 검정하기 때문에 `전체 1종 오류`가 증가할 수 있다.

2) 다중 비교 문제의 발생 원인 및 예시

    - 발생 원인: 여러 번의 독립적인 가설 검정을 수행할 때, `각 검정이 독립적이지 않다면`, 전체 오류가 누적된다.
    
        - 이는 여러 비교를 통해 유의하지 않은 결과를 유의미하다고 오인하게 한다.
    
    - 예시 1: 𝑛개의 그룹 간 평균 차이를 비교하기 위해 𝑛(𝑛−1)/2개의 t-검정을 수행하면,
    
        - 유의수준 𝛼=0.05로 각 검정을 독립적으로 진행하더라도 전체 오류가 크게 증가한다.

    - 예시 2: 20개의 약을 각각의 효과를 검정할 때, 각각의 검정에서 유의수준을 0.05로 설정했다면,
    
        - 단일 검정의 오류율은 5%이지만, 20개의 검정 전체에서 오류가 발생할 확률은 64%에 이른다.

        - 즉, 20개의 가설을 검정할 때, 약 64%의 확률로 하나 이상의 잘못된 기각이 발생할 수 있음을 의미한다.

```python
# 연구자가 20개의 가설을 검정할 때, 약 64%의 확률로 하나 이상의 잘못된 기각이 발생할 수 있다.
import numpy as np

# 예시: 20개의 가설 검정, 각 검정에서 유의수준이 0.05인 경우
n_tests = 20
alpha = 0.05

# 하나 이상의 1종 오류가 발생할 확률(FWER)
prob_one_or_more_errors = 1 - (1 - alpha) ** n_tests

print(f"하나 이상의 1종 오류가 발생할 확률: {prob_one_or_more_errors:.2%}")
```

```python
# 연구자가 10개의 다른 치료법을 비교하면서 45개의 가설 검정을 수행한다면, 이 중 하나 이상의 잘못된 유의미한 결과가 나올 확률이 매우 높다.
import numpy as np

# 예시: 10개의 치료법 비교 시 발생하는 1종 오류 증가
n_tests = 45
alpha = 0.05

# 하나 이상의 1종 오류가 발생할 확률 계산
prob_one_or_more_errors = 1 - (1 - alpha) ** n_tests
print(f"하나 이상의 1종 오류 발생 확률: {prob_one_or_more_errors:.2%}")
```

3) 다중 비교 문제의 중요성 및 영향

    - 다중 비교 문제는 연구 결과의 신뢰도를 심각하게 저하시킬 수 있다.
    
    - 잘못된 양성 결과가 증가하면, 실제로는 유의미하지 않은 차이를 유의미하다고 판단할 가능성이 높아지기 때문에, 이를 해결하기 위한 통계적 방법이 필수적이다.

### 2. 다중 비교의 유형

1) 단일 가설 검정과 다중 가설 검정의 차이

    - 단일 가설 검정에서는 하나의 가설만을 대상으로 하므로, 유의수준을 쉽게 관리할 수 있다.
    
    - 그러나 다중 가설 검정에서는 여러 가설을 동시에 검정하기 때문에, 각각의 검정이 `독립적이지 않으면` 전체 유의수준을 초과할 수 있다.

    - 이는 각각의 검정이 독립적이지 않으면 전체 오류율이 상승한다는 말과 같다.

2) 후속 분석에서의 다중 비교

    - 연구에서 초기 분석 후 추가적인 가설 검정을 수행하는 경우도 다중 비교 문제를 일으킬 수 있다.
    
    - 예를 들어, 그룹 간의 차이를 조사하기 위해 여러 가설을 순차적으로 검정하면 다중 비교 문제가 발생할 수 있다.

3) 탐색적 데이터 분석(EDA)에서의 다중 비교

    - 탐색적 데이터 분석에서 여러 패턴이나 관계를 발견하려는 과정에서 여러 가설 검정이 수행될 수 있으며, 이로 인해 다중 비교 문제가 발생할 수 있다.

    - EDA의 목표는 데이터에서 패턴을 발견하는 것이지만, 이 과정에서 오류율이 증가할 수 있다.
    
    - EDA에서 특히 주의가 필요하다.

### 3. 다중 비교로 인한 오류

1) 1종 오류(허위 양성)의 증가

    - 다중 비교를 할 때 1종 오류가 증가하는데, 이는 귀무가설이 참일 때도 귀무가설을 기각하는 오류를 말한다.
    
    - 다중 비교 문제는 이 오류를 크게 증가시킨다.

2) 2종 오류(허위 음성)와의 관계

    - 2종 오류는 귀무가설이 거짓일 때 이를 기각하지 못하는 오류이다.
    
    - 다중 비교 문제를 해결하기 위해 보수적인 방법을 사용할 경우 1종 오류를 줄일 수 있지만, 2종 오류가 증가할 수 있다.

3) 가족/그룹/집합별 오류율(FWER, Family-Wise Error Rate)

    - FWER는 특정 집합(예: 다중 비교의 전체 검정 집합)에서 하나 이상의 1종 오류(잘못된 기각)가 발생할 확률을 의미한다.

    - 이를 제어하는 방법이 다중 비교 문제의 핵심⭐이다.

    - $FWER = 1 - (1 - \alpha)^m$ 와 같이 계산되며, $\alpha$는 각 검정의 유의수준, $m$은 검정의 수이다.
    
    - 다중 비교에서 FWER을 제어하기 위한 다양한 방법들이 개발되었다.

4) 유의수준의 관리 및 희석 효과

    - 다중 비교를 수행할 때 전체 유의수준을 관리하지 않으면, 개별 검정에서 설정한 유의수준보다 전체 유의수준이 희석될 수 있다.

    - 즉, 각 검정의 유의수준을 전체 검정의 유의수준으로 조정하지 않으면, 전체 유의수준이 희석될 수 있다.

    - 희석 효과는 여러 번의 검정으로 인해 본래의 효과가 약해지는 현상을 의미한다.
    
    - 이는 다중 비교 문제에서 유의한 결과를 찾기 어렵게 만든다.

### 4. 다중 비교 문제의 해결 방법

1) 보수적 방법과 자유도 감소의 영향

    - Bonferroni 교정처럼 보수적인 방법은 1종 오류를 줄이는데 효과적이지만,
    
    - 그만큼 `검정력`이 떨어지며, `자유도`가 감소해 2종 오류가 증가할 수 있다.

2) Bonferroni 교정

    - Bonferroni 교정은 가장 간단하고 보수적인 방법 중 하나이다.
    
    - 개별 가설 검정에서 사용된 유의수준(α)을 가설의 수만큼 나누어 전체 오류(Family-Wise Error Rate)를 제어하는 방법이다.
    
    - 즉, 수정된 유의수준 𝛼′=𝛼/𝑚(m은 가설의 수)로 각 검정을 수행한다.

### ⭐tip.자유도와 검정력의 상관관계

1. 자유도와 검정력

    - 자유도(Degrees of Freedom)는 통계 검정에서 계산에 사용되는 독립적인 데이터 포인트의 수를 나타낸다.
    
    - 자유도가 높다는 것은 더 많은 정보를 기반으로 검정을 수행한다는 의미이다.

    - 검정력(Power)은 실제로 효과가 존재할 때 이를 발견할 수 있는 능력이다.
    
    - 검정력은 주로 1종 오류와 2종 오류의 균형을 맞추는 역할을 하며, 자유도가 검정력에 중요한 영향을 미친다.

2. 관계

    - 자유도가 증가할수록 검정력도 증가한다.
    
        - 이는 더 많은 데이터나 더 복잡한 모델이 검정력을 높여, 실제 효과를 더 잘 감지할 수 있기 때문이다.

    - 자유도가 낮아지면 검정력도 감소할 수 있다.
    
        - 이는 정보가 부족하거나 비교하는 그룹의 수가 많아질 때 발생할 수 있으며, 효과를 감지하기 어려워지므로 2종 오류가 증가할 가능성이 크다.

3. 예시

    - 분산분석(ANOVA)에서 자유도는 표본 크기와 관련된다.
    
    - 표본 크기가 커지면 자유도가 높아지고, 이로 인해 검정력도 상승하여 미세한 차이도 더 쉽게 감지할 수 있다.

4. 결론

    - 검정력과 자유도는 양의 상관관계가 있다.
    
    - 자유도가 높아지면 검정력이 증가하여, 실제로 효과가 있을 때 이를 감지할 가능성이 커진다.
    
    - 반대로, 자유도가 낮아지면 검정력이 떨어져 효과를 발견하기 어려워지며, 이로 인해 2종 오류가 증가할 수 있다.

```python
alpha = 0.05  # 개별 가설 검정에서 사용된 유의수준
m = 20  # 가설의 수

# Bonferroni 교정된 유의수준
alpha_corrected = alpha / m
print(f"Bonferroni 교정된 유의수준: {alpha_corrected:.4f}")
```

3) Holm-Bonferroni 방법

    - Holm-Bonferroni 방법은 Bonferroni 교정보다 덜 보수적인 방법이다.
    
    - p-값들을 오름차순으로 정렬한 후, 가장 작은 p-값부터 점진적으로 유의수준을 조정하여 검정한다.

    - 이 방법은 Bonferroni 교정보다 더 강력한 방법이다.

4) Benjamini-Hochberg 절차 (FDR: False Discovery Rate)

    - Benjamini-Hochberg 절차는 FWER 대신 FDR(허위 발견율)을 제어하는 방법이다.
    
    - 이는 다중 비교에서 검정력의 손실 없이 1종 오류를 제어할 수 있도록 설계되었다.

    - 즉, 유의미한 결과를 더 많이 발견하면서도 1종 오류를 효과적으로 제어한다.

    - 방법: p-value를 정렬한 후, 각 p-value에 대해 다음과 같은 기준을 적용한다.

        - $p_i \leq \frac{i}{m} \cdot Q$

        - $p_i$는 i번째 p-value, $m$은 총 검정 수, $Q$는 원하는 FDR이다.

5) Scheffé 및 Tukey의 사후 검정

    - Scheffé와 Tukey의 사후 검정은 분산분석(ANOVA) 후에 어떤 그룹 간에 차이가 있는지(다중 비교) 확인하는 방법이다.

    - Scheffé 검정: 모든 가능한 비교를 고려하는 보수적인 사후 검정 방법이다. 더 일반적인 비교를 수행한다.

    - Tukey의 HSD: 모든 그룹 쌍의 평균 차이 비교하는 방법이다.

6) Hochberg's GT2, Gabriel’s, Games-Howell 방법 등

    - Hochberg's GT2: 표본 크기가 다른 경우에 적합한 다중 비교 방법이다.

    - Games-Howell 방법: 분산이 동일하지 않은 그룹 간 비교에 적합한 방법이다.

### 5. 다중 비교에서의 데이터 시각화

1) 주요 다중 비교 방법의 시각적 표현

    - 다중 비교 결과를 이해하고 전달하기 위해 시각화를 활용할 수 있다.
    
    - 예를 들어, Tukey HSD 결과는 그룹 간 평균 차이를 시각적으로 보여줄 수 있다.

        - Tukey HSD(Honestly Significant Difference)는 여러 그룹 간의 평균 차이를 비교할 때 사용되는 사후 검정 방법이다.
        
        - Tukey HSD 결과를 시각적으로 표현하는 방법으로는 막대 그래프와 신뢰 구간을 포함한 박스플롯 등이 있다.

        - 아래는 Python의 statsmodels 라이브러리를 사용하여 Tukey HSD 결과를 시각화하는 코드 예제이다.
        
        - 이 예제에서는 statsmodels를 이용해 Tukey HSD 검정을 수행하고, 결과를 시각적으로 표시한다.

    - 코드를 설명하자면 다음과 같다.

        - 데이터 생성: np.random.normal을 사용해 3개의 그룹(A, B, C)의 값을 생성한다.

        - ANOVA 모델 적합: statsmodels의 ols를 사용하여 ANOVA 모델을 적합한다.

        - Tukey HSD 검정: pairwise_tukeyhsd를 사용하여 Tukey HSD 검정을 수행한다.

        - 박스플롯 시각화: seaborn의 boxplot을 사용하여 각 그룹의 값을 시각적으로 비교한다.

        - Tukey HSD 결과 시각화: Tukey HSD 결과를 plot_simultaneous를 사용하여 시각화한다. 이 그래프는 그룹 간의 평균 차이를 나타내며, 신뢰 구간을 포함하여 비교한다.

2) 결과 해석 및 시각화 시 유의할 점

    - 다중 비교 결과를 시각화할 때는 `오류율`을 명확히 표시해야 하며, 신뢰 구간 등을 통해 해석의 신뢰성을 높여야 한다.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 예제 데이터 생성
np.random.seed(0)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(loc, 10, 30) for loc in range(3)]),
    'group': ['A']*30 + ['B']*30 + ['C']*30
})

# 데이터 확인
print(data.head())

# ANOVA 모델 적합
model = ols('value ~ C(group)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Tukey HSD 검정
tukey = pairwise_tukeyhsd(endog=data['value'], groups=data['group'], alpha=0.05)
print(tukey)

# Tukey HSD 결과 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='group', y='value', data=data)
plt.title('Group-wise Boxplot with Tukey HSD Results')
plt.show()

# Tukey HSD 결과 시각화 (결과 테이블)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
tukey.plot_simultaneous(ax=ax)
plt.title('Tukey HSD Results')
plt.show()
```

### ⭐tip.그래프 해석 방법

1. 박스플롯 (Boxplot)

    - 박스플롯을 사용하여 그룹 간의 중앙값, 분포, 그리고 이상치 등을 비교한다.

    - 그룹 간의 중앙값 차이가 클수록 두 그룹 간의 평균 차이가 클 가능성이 있다.

    - 그룹 간의 상자와 수염의 위치가 겹치지 않으면, Tukey HSD 검정에서 유의미한 차이가 있을 가능성이 있다.

2. Tukey HSD 결과 시각화 (Tukey HSD Plot)

    - Tukey HSD 결과 시각화는 Tukey HSD 검정의 결과를 직관적으로 보여준다.
    
    - 주로 다음과 같은 요소가 포함된다.

        1) 중앙 점 (Mean)
        
            - 각 그룹의 평균값으로, Tukey HSD 그래프에서 각 그룹 간의 평균 차이를 비교할 수 있다.

        2) 신뢰 구간 (Confidence Intervals)
        
            - 중앙 점의 좌우에 있는 수평선으로, 평균 차이의 신뢰 구간을 나타낸다.
            
            - 신뢰 구간이 0을 포함하지 않으면, 그룹 간의 평균 차이는 통계적으로 유의미하다.

    - 해석 방법

        1) 평균 차이: 그래프에서 그룹 간의 평균 차이를 시각적으로 비교한다. 평균 차이가 크면 그룹 간의 차이가 더 명확하다는 것을 의미한다.

        2) 신뢰 구간: 신뢰 구간이 0을 포함하지 않으면, 두 그룹 간의 평균 차이가 통계적으로 유의미하다고 판단한다. 신뢰 구간이 0을 포함하면, 차이가 유의미하지 않을 수 있다.
        
        3) 간격: 신뢰 구간의 길이가 짧을수록 평균 차이의 추정이 더 정확하다고 할 수 있다.

    - 해석 결과

        1) 신뢰 구간이 0을 포함하는 경우

            - B와 C그룹 간의 평균 차이를 나타내는 신뢰 구간이 0을 포함하면, 두 그룹 간의 평균 차이가 유의미하지 않다는 뜻이다.
            
                - 이는 두 그룹 간의 차이가 우연히 발생할 수 있는 정도일 수 있으며,
                
                - 두 그룹의 평균이 실제로 유의미하게 다르지 않을 가능성을 시사한다.

            - B와 C그룹은 서로 비교했을 때 평균 차이가 통계적으로 유의미하지 않다고 판단된다.

        2) 신뢰 구간이 0을 포함하지 않는 경우

            - 반면, 신뢰 구간이 0을 포함하지 않으면, 두 그룹 간의 평균 차이가 통계적으로 유의미하다는 것을 의미한다.
            
            - 즉, 두 그룹의 평균이 실제로 다를 가능성이 높다.

            - A그룹은 B와 C그룹과 비교했을 때 평균 차이가 통계적으로 유의미하다.

### ⭐tip.오류율 (Error Rate)

1. 정의
    
    - 오류율은 특정 오류가 발생할 확률로, 여러 가지 종류가 있다.

2. 예시

    - 1종 오류율: 유의수준과 동일하며, 잘못된 기각의 확률이다.

    - FWER(Family-Wise Error Rate): 다중 비교 시 하나 이상의 1종 오류가 발생할 확률이다.

    - FDR (False Discovery Rate, 거짓 발견율): 기각된 가설 중에서 실제로 거짓인 가설의 비율이다.

3. 유의수준과 차이점 요약

    - 유의수준은 개별 가설 검정의 1종 오류율을 나타내며, 다중 비교 문제에서는 전체 오류율(FWER)을 고려해야 한다.
    
    - 다중 비교에서는 각 가설의 유의수준을 낮추더라도, 전체 오류율이 상승할 수 있기 때문에, 이를 명확히 표시하고 관리해야 한다.

### ⭐tip.FWER과 FDR 비교

1. FWER (Family-Wise Error Rate, 가족별 오류율)

    1) 정의

        - FWER는 다중 비교를 수행할 때, 하나 이상의 1종 오류(허위 양성)가 발생할 확률을 나타낸다.
    
        - 즉, 귀무가설이 참일 때 이를 잘못 기각할 확률의 총합이다.

    2) 설명

        - FWER는 전체 검정에서 적어도 하나의 1종 오류가 발생할 확률을 측정한다.
        
        - 이 확률이 높으면 여러 검정을 수행할 때 1종 오류가 발생할 가능성이 커지므로, 다중 비교 문제를 다루는 데 중요히다.

    3) 계산 예시

        - 5개의 검정을 수행하고 각 검정의 유의수준을 0.05로 설정한 경우, FWER는 다음과 같이 계산할 수 있다.

        - 

            - 𝛼 : 개별 유의수준
        
            - 𝑚 : 가설 검정의 수
        
        - 예를 들어, 𝛼=0.05, 𝑚=5일 때, FWER는 약 0.23이다.

2. FDR (False Discovery Rate, 거짓 발견율)

    1) 정의

        - FDR은 기각된 가설 중에서 실제로 거짓인 가설의 비율을 나타낸다.
        
        - 즉, 여러 검정에서 유의미하다고 판별된 결과 중 실제로는 유의미하지 않은 결과의 비율이다.

    2) 설명

        - FDR은 다중 비교에서 허용 가능한 허위 발견의 비율을 제어하는 데 중점을 둔다.
        
        - Benjamini-Hochberg 절차와 같은 방법이 FDR을 제어하는 데 사용된다.

    3) 계산 예시

        - FDR은 다음과 같이 계산할 수 있다.

        - 
    
            - "False Positives" : 잘못된 유의미한 결과의 수
        
            - "Total Rejected Hypotheses" : 기각된 가설의 총 수

3. 요약

    - 두 개념은 다중 비교에서의 오류를 제어하기 위해 사용된다.
    
    - FWER는 전체적인 오류 발생 확률을 줄이기 위한 방법이고, FDR은 기각된 가설 중 실제로 거짓인 비율을 제어한다.

### 6. 다중 비교 문제와 데이터 분석 사례

1) 다중 비교 문제를 고려한 실제 데이터 분석 사례

    - 여러 그룹 간의 차이를 비교하는 연구에서, 다중 비교 문제를 고려하여 Bonferroni 교정을 적용한 사례를 찾아본다.

    - 예를 들어, 다양한 치료법의 효과를 비교할 때 각 그룹 간의 차이를 다중 비교 방법을 사용해 분석할 수 있다.

2) 다중 비교 문제 무시 시 발생할 수 있는 오류 사례

    - 다중 비교 문제를 무시하고 단순히 p-value를 비교할 경우, 잘못된 결론에 도달할 수 있다.
    
    - 예를 들어, 0.05보다 작은 p-value가 여러 개 존재할 경우, 실제로는 차이가 없는 경우에도 차이가 있다고 결론지을 수 있다.

    - 때문에 통계적으로 유의하지 않은 차이를 유의하다고 판단하는 오류(허위 양성)가 발생할 수 있다.

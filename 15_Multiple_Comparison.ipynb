        - ì•„ëž˜ëŠ” Pythonì˜ statsmodels ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ Tukey HSD ê²°ê³¼ë¥¼ ì‹œê°í™”í•˜ëŠ” ì½”ë“œ ì˜ˆì œì´ë‹¤.
        
        - ì´ ì˜ˆì œì—ì„œëŠ” statsmodelsë¥¼ ì´ìš©í•´ Tukey HSD ê²€ì •ì„ ìˆ˜í–‰í•˜ê³ , ê²°ê³¼ë¥¼ ì‹œê°ì ìœ¼ë¡œ í‘œì‹œí•œë‹¤.

        - ë°ì´í„° ìƒì„±: np.random.normalì„ ì‚¬ìš©í•´ 3ê°œì˜ ê·¸ë£¹(A, B, C)ì˜ ê°’ì„ ìƒì„±í•œë‹¤.

        - ANOVA ëª¨ë¸ ì í•©: statsmodelsì˜ olsë¥¼ ì‚¬ìš©í•˜ì—¬ ANOVA ëª¨ë¸ì„ ì í•©í•œë‹¤.

        - Tukey HSD ê²€ì •: pairwise_tukeyhsdë¥¼ ì‚¬ìš©í•˜ì—¬ Tukey HSD ê²€ì •ì„ ìˆ˜í–‰í•œë‹¤.

        - ë°•ìŠ¤í”Œë¡¯ ì‹œê°í™”: seabornì˜ boxplotì„ ì‚¬ìš©í•˜ì—¬ ê° ê·¸ë£¹ì˜ ê°’ì„ ì‹œê°ì ìœ¼ë¡œ ë¹„êµí•œë‹¤.

        - Tukey HSD ê²°ê³¼ ì‹œê°í™”: Tukey HSD ê²°ê³¼ë¥¼ plot_simultaneousë¥¼ ì‚¬ìš©í•˜ì—¬ ì‹œê°í™”í•œë‹¤. ì´ ê·¸ëž˜í”„ëŠ” ê·¸ë£¹ ê°„ì˜ í‰ê·  ì°¨ì´ë¥¼ ë‚˜íƒ€ë‚´ë©°, ì‹ ë¢° êµ¬ê°„ì„ í¬í•¨í•˜ì—¬ ë¹„êµí•œë‹¤.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# ì˜ˆì œ ë°ì´í„° ìƒì„±
np.random.seed(0)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(loc, 10, 30) for loc in range(3)]),
    'group': ['A']*30 + ['B']*30 + ['C']*30
})

# ë°ì´í„° í™•ì¸
print(data.head())

# ANOVA ëª¨ë¸ ì í•©
model = ols('value ~ C(group)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Tukey HSD ê²€ì •
tukey = pairwise_tukeyhsd(endog=data['value'], groups=data['group'], alpha=0.05)
print(tukey)

# Tukey HSD ê²°ê³¼ ì‹œê°í™”
plt.figure(figsize=(10, 6))
sns.boxplot(x='group', y='value', data=data)
plt.title('Group-wise Boxplot with Tukey HSD Results')
plt.show()

# Tukey HSD ê²°ê³¼ ì‹œê°í™” (ê²°ê³¼ í…Œì´ë¸”)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
tukey.plot_simultaneous(ax=ax)
plt.title('Tukey HSD Results')
plt.show()
```

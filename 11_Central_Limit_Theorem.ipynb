### 1. 중심극한정리의 개요

1) 중심극한정리(Central Limit Theorem, CLT)란 무엇인가?

    - 통계학의 기본 이론으로, 모집단의 분포가 어떠하든, 표본의 크기가 충분히 크다면 표본 평균의 분포가 정규 분포에 가까워진다는 이론이다.
    
    - = 모집단이 비정규분포일지라도 표본의 크기가 증가하면 표본 평균의 분포는 정규분포에 가까워진다.

    - 이로 인해 정규 분포가 아닌 데이터라도 표본 평균을 이용해 통계적 추정을 할 수 있다.

    - = 통계적 추정을 할 때 표본 평균을 사용하여 모집단의 특성을 파악할 수 있도록 해준다.

2) 중심극한정리의 역사적 배경과 중요성

    - 중심극한정리는 18세기 후반에 Pierre-Simon Laplace와 Carl Friedrich Gauss에 의해 발전되었다.
    
    - 이 이론은 통계학과 확률론의 기초를 형성하며, 다양한 분야에서의 데이터 분석과 가설 검정, 신뢰 구간 추정 등에서 필수적으로 사용된다.
    
    - 특히, 복잡한 데이터를 분석할 때 정규 분포의 특성을 활용할 수 있게 해주는 중요한 도구가 된다.

    - 그 중요성은 통계적 방법의 기초가 되기 때문에, 실험 결과나 조사 데이터를 해석하는 데 매우 유용하다.

### 2. 중심극한정리의 이론적 기초

1) 모집단과 표본 분포

    - 모집단(Population): 연구 대상 전체 집합을 의미한다. 예를 들어, 한 나라의 모든 성인 남성을 모집단으로 설정할 수 있다.

    - 표본(Sample): 모집단에서 선택된 일부 데이터를 의미한다. 표본은 모집단의 특성을 추정하는 데 사용된다.

    - 표본 분포(Sampling Distribution): 특정 통계량(예: 표본 평균)의 분포를 의미한다. 표본 평균의 분포는 모집단의 분포에 따라 달라진다.
    
    - 모집단 분포가 비정규 분포라도, 중심극한정리 덕분에 표본 평균의 분포가 정규 분포를 따르게 된다.

2) 확률 변수와 독립 동일 분포

    - 확률 변수(Probability Variable): (불확실한) 실험의 결과를 수치로 나타낸 것이다.
    
        - 예를 들어, 동전을 던져서 앞면이 나오는 경우를 1, 뒷면이 나오는 경우를 0으로 나타낼 수 있다.

    - 독립 동일 분포(Independent and Identically Distributed, i.i.d): 각 표본이 동일한 분포를 따르면서 서로 독립적이라는 의미이다.
    
        - 즉, 각 표본의 선택이 다른 표본의 선택에 영향을 미치지 않는다.
    
    - 중심극한정리는 이 두 조건이 성립할 때 적용된다.

3) 표본 평균과 그 분포

    - 표본 평균(Sample Mean): 주어진 표본의 모든 관측값을 더한 후 표본의 크기로 나눈 값이다.

        - 표본 평균은 표본의 크기($𝑛$)가 커질수록 모집단 평균에 가까워진다.
    
        - 표본 평균은 모집단의 평균을 추정하는 데 사용된다.

    - 표본 평균의 분포: 표본 평균은 반복적으로 표본을 추출할 때 생성되는 평균값의 분포를 의미한다.
    
        - 중심극한정리에 따르면, 표본 평균의 분포는 표본 크기가 커질수록 정규분포에 가까워진다.

### 3. 중심극한정리의 수학적 정의

1) 중심극한정리의 수식 표현

    - 중심극한정리는 다음과 같이 수식으로 표현할 수 있다.

        - 표본 크기를 $n$이라고 할 때, 모집단의 평균을 $\mu$, 모집단의 분산을 $\sigma^2$라고 하면?
    
        - $n$이 충분히 클 경우 표본 평균 $\bar{X}$의 분포는 다음과 같다.

        - $\bar{X} \sim N\left(\mu, \frac{\sigma^2}{n}\right)$

            - $\bar{X}$ : 표본 평균

            - $N(\mu, \frac{\sigma^2}{n})$ : 평균 $\mu$와 분산 $\frac{\sigma^2}{n}$인 정규분포
                
                - $𝜇$ :모집단 평균
                
                - $𝜎$ : 모집단 표준편차
                
                - $𝑛$ : 표본 크기

2) 표본 크기와 표준 오차의 관계

    - 표준 오차(Standard Error, SE): 표본 평균의 분포의 표준편차를 의미한다. 이는 표본 크기 $n$에 따라 다음과 같이 정의된다.

        - $SE = \frac{\sigma}{\sqrt{n}}$

    - 표본 크기 $𝑛$이 증가할수록 표준 오차 $SE$는 감소하며, 이는 표본 평균이 모집단 평균에 더 가까워짐을 의미한다.
    
    - 즉, 표본 크기가 커질수록 표본 평균이 모집단 평균에 더욱 근접하게 된다.

3) 대수의 법칙(Law of Large Numbers)과의 관계

    - 대수의 법칙은 표본의 크기가 커질수록 표본 평균이 모집단 평균에 수렴한다는 법칙이다.
    
    - 이는 중심극한정리와 함께 사용할 수 있으며, 대수의 법칙은 표본 크기가 클수록 더 정확한 추정치를 제공한다는 점에서 중심극한정리의 기초가 된다.

    - 중심극한정리와 유사하지만, `대수의 법칙`은 표본 평균이 하나의 값으로 수렴하는 반면, `중심극한정리`는 표본 평균의 분포가 정규 분포에 수렴한다는 점에서 차이가 있다.

### 4. 중심극한정리의 시각적 이해

1) 다양한 분포에서의 중심극한정리 시뮬레이션

    - 여러 가지 분포(예: 균등분포, 지수분포)에서 표본 평균의 분포가 어떻게 정규분포에 근접하는지를 예제를 통해 확인할 수 있다.

    - 중심극한정리를 시각적으로 이해하기 위해, 비정규 분포에서 추출한 표본 평균의 분포를 확인해보자.

    - 아래 코드는 비대칭적인 지수 분포를 사용하여 표본 평균의 분포가 어떻게 정규 분포에 수렴하는지를 보여준다.

```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt

# 모집단: 비대칭 지수 분포
population = np.random.exponential(scale=2, size=100000)

# 표본 크기 별로 표본 평균의 분포 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sample_sizes = [10, 30, 100]

for i, size in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(population, size=size)) for _ in range(1000)]
    axes[i].hist(sample_means, bins=30, alpha=0.7, color='blue')
    axes[i].set_title(f'Sample Size = {size}')
    axes[i].set_xlabel('Sample Mean')
    axes[i].set_ylabel('Frequency')

plt.show()
```

```python
# 필요한 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 설정
np.random.seed(0)
sample_size = 30  # 표본 크기
num_samples = 1000  # 생성할 표본 수

# 비정규 분포 (예: 균등분포)
data = np.random.uniform(low=0, high=100, size=num_samples)

# 표본 평균 계산
sample_means = [np.mean(np.random.choice(data, sample_size)) for _ in range(num_samples)]

# 히스토그램 그리기
plt.figure(figsize=(8, 4))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
plt.title('표본 평균의 분포 (균등분포에서 추출)')
plt.xlabel('표본 평균')
plt.ylabel('밀도')
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=2)
plt.show()
```

2) 표본 크기에 따른 분포의 변화

    - 위의 예제를 통해 알 수 있듯이, 표본 크기가 커질수록 표본 평균의 분포는 점점 더 정규 분포에 가까워진다.

    - 이는 중심극한정리를 시각적으로 이해하는 데 도움을 준다.

### 5. 중심극한정리의 조건

1. 독립성 (Independence)

    - 표본들이 서로 독립적이어야 한다.
    
    - 즉, 하나의 표본이 다른 표본에 영향을 미치지 않아야 한다.
    
    - 예를 들어, 동전 던지기에서 한 번의 던지기 결과가 다음 던지기 결과에 영향을 주지 않는 것처럼, 각각의 표본은 서로 독립적이어야 한다.

2. 동일 분포 (Identical Distribution)

    - 모든 표본은 동일한 확률 분포를 따라야 한다.
    
    - 즉, 각 표본이 같은 분포에서 추출되어야 하며, 이는 모집단의 특성을 반영해야 한다.
    
    - 예를 들어, 한 그룹의 성인 남성을 모집단으로 할 때, 모든 성인이 동등한 확률로 선택되어야 한다.

3. 충분히 큰 표본 크기

    - 표본의 크기가 충분히 커야 한다. 일반적으로 표본 크기 𝑛이 30 이상이면 중심극한정리가 성립된다고 본다.
    
    - 그러나 모집단의 분포가 매우 비대칭이거나 극단적인 값을 포함하는 경우에는 더 큰 표본 크기가 필요할 수 있다.
    
    - 표본 크기가 클수록 표본 평균의 분포는 정규분포에 가까워진다.

### 6. 중심극한정리의 실제 적용

1. 통계적 추정과 중심극한정리

    - 중심극한정리는 통계적 추정의 기초가 된다.
    
    - 예를 들어, 모집단의 평균을 추정할 때, 표본에서 얻은 표본 평균을 사용하여 모집단 평균을 추정할 수 있다.
    
    - 표본 평균이 정규분포를 따른다는 것을 알기 때문에, 이를 통해 `신뢰 구간`을 계산하거나 `가설 검정`을 할 때 매우 유용하다.

2. `가설 검정`에서의 활용

    - 가설 검정에서는 중심극한정리를 통해 표본 평균의 분포를 정규 분포로 가정하고, 이를 바탕으로 통계적 유의성을 평가한다.

    - 예를 들어, 특정 약물의 효과를 검증하기 위해 두 집단의 평균을 비교하는 경우, 표본 평균이 정규분포를 따른다는 사실을 이용하여 z-검정이나 t-검정을 수행할 수 있다.
    
    - 이는 귀무가설을 검정하는 데 필수적인 요소이다.

3. `신뢰 구간` 추정에서의 역할

    - 신뢰 구간을 추정할 때, 중심극한정리는 표본 평균이 정규 분포를 따른다고 가정하여 신뢰 구간을 계산할 수 있게 해준다.

    - 예를 들어, 모집단 평균에 대한 95% 신뢰 구간을 설정할 때, 표본 평균과 표준 오차를 이용하여 다음과 같이 계산한다.

        - $\text{신뢰 구간} = \bar{X} \pm z_{\alpha/2} \times SE$

        - 여기서 $z_{\alpha/2}$는 정규분포에서의 임계값이다.

### 7. 중심극한정리의 한계

1. 표본 크기가 작은 경우의 문제점

    - 표본 크기가 작을 경우, 표본 평균이 정규분포를 따르지 않을 수 있다. 이는 표본이 모집단의 특성을 잘 반영하지 못하기 때문이다.
    
    - 따라서 표본 크기가 작을 경우, 중심극한정리를 적용하기 어렵다.

    - 이 경우 정규 분포를 가정하는 것이 잘못된 결과를 초래할 수 있다.

2. 비대칭 분포에서의 적용 제한

    - 모집단 분포가 매우 비대칭적이거나 극단적인 경우, 표본 크기가 커야만 중심극한정리가 성립합니다.

    - 예를 들어, 소득 분포와 같이 극단적인 값을 포함한 분포는 비대칭적이며, 이러한 경우에는 더 큰 표본 크기가 필요하다.

3. 모집단의 분산이 무한대일 경우

    - 모집단의 분산이 무한대인 경우, 중심극한정리가 성립하지 않는다.

    - 예를 들어, 파레토 분포와 같은 경우는 분산이 정의되지 않기 때문에 표본 평균의 분포가 정규분포에 수렴하지 않는다.

    - 이 경우 다른 방법을 고려해야 한다.

### 8. 중심극한정리와 데이터 분석

1. 실무에서의 중심극한정리 적용 사례

    - 실무에서는 중심극한정리를 바탕으로 여러 통계적 방법을 적용한다.
    
    - 예를 들어, A/B 테스트에서 표본 평균의 차이를 분석할 때, 중심극한정리가 활용된다.

    - 또는 마케팅 조사에서 소비자 설문을 통해 평균 소비액을 추정하거나, 의료 연구에서 약물의 효과를 평가할 때 사용된다.
    
    - 이러한 적용은 데이터 분석의 신뢰성을 높이는 데 기여한다.

2. 실험 데이터와 관측 데이터에서의 차이

    - 실험 데이터는 연구자가 조작한 변수를 기반으로 수집된 데이터이며, 관측 데이터는 자연적인 상황에서 수집된 데이터이다.

    - 예를 들어, 실험 데이터에서는 처치 효과를 검증할 때, 관측 데이터에서는 인과관계를 분석할 때 활용된다.

    - 실험 데이터는 일반적으로 잘 통제된 환경에서 얻어지기 때문에 중심극한정리가 잘 적용되지만, 관측 데이터는 그렇지 않은 경우도 있다.
    
    - 이를 고려한 추가적인 분석이 필요하다.
    
3. 머신러닝 모델에서의 활용

    - 머신러닝에서도 중심극한정리는 중요한 역할을 한다.
    
    - 예를 들어, 모델의 성능 평가를 위해 교차 검증을 수행할 때, 각 fold의 성능 지표의 평균을 계산하는 과정에서 중심극한정리를 활용할 수 있다.
    
    - 이때, 여러 번의 평가를 통해 얻은 평균 성능 지표는 정규분포를 따르게 된다.

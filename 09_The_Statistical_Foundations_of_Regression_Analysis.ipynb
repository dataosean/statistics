### 1. 회귀분석의 개요

1) 회귀분석의 정의 및 목적

    - 회귀분석은 두 개 이상의 변수 간의 관계를 분석하는 통계 기법이다.
    
    - 주로 하나의 종속 변수(결과 변수)를 여러 독립 변수(설명 변수)를 사용해 예측하거나 설명할 때 사용된다.
    
    - 목적은 주어진 데이터로부터 회귀식을 도출해 미래의 값을 예측하거나 변수들 간의 관계를 파악하는 것이다.

2) 회귀분석의 유형 (단순회귀 vs 다중회귀)

    - 단순회귀분석: 하나의 독립 변수와 하나의 종속 변수 간의 관계를 분석한다.

        - $Y = \beta_0 + \beta_1 X + \epsilon$

            - $Y$: 종속 변수
            
            - $X$: 독립 변수
            
            - $\beta_0$: 절편(Y축과 만나는 지점)
            
            - $\beta_1$: 기울기(X가 1 단위 증가할 때 Y의 변화량)
            
            - $\epsilon$: 오차항
    
        - 예를 들어, '공부 시간'이 '시험 성적'에 미치는 영향을 분석하는 경우가 해당된다.

    - 다중회귀분석: 두 개 이상의 독립 변수를 이용해 하나의 종속 변수를 예측한다.

        - $Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + ... + \beta_n X_n + \epsilon$

            - 여기서 $X_1, X_2, ..., X_n$은 여러 독립 변수이다.
    
        - 예를 들어, '공부 시간', '수면 시간', '참고서 활용'이 '시험 성적'에 미치는 영향을 동시에 분석하는 경우가 해당된다.

3) 회귀분석의 응용 분야

    - 회귀분석은 경제학, 생물학, 공학, 마케팅 등 다양한 분야에서 사용된다.

        - 경제학: 소비자 행동 분석, 가격 예측한다.
        
        - 생물학: 생물 종의 생존율 예측한다.

        - 사회과학: 설문조사 데이터를 통한 사회적 요인 분석한다.
    
        - 마케팅: 광고 비용이 매출에 미치는 영향을 분석할 때 회귀분석을 사용한다.

### 2. 선형 회귀분석의 기본 개념

1)  독립 변수와 종속 변수

    - 독립 변수(Independent Variable)
        
        - 결과에 영향을 미치는 변수. 예를 들어, 집값을 예측할 때는 '평수'가 독립 변수이다.

    - 종속 변수(Dependent Variable)
        
        - 예측하고자 하는 변수. 예를 들어, '집값'이 종속 변수이다.

2) 선형 관계와 회귀선의 의미

    - 선형 회귀는 독립 변수와 종속 변수 사이의 선형 관계를 가정한다.
    
    - 선형 관계는 변수 간의 관계가 직선으로 나타나는 것을 의미한다.
    
    - 회귀선은 종속 변수와 독립 변수 간의 관계를 시각적으로 나타내는 선으로, 이를 통해 예측을 수행한다.
    
        - 회귀선은 데이터 포인트들을 가장 잘 설명하는 직선이라고 할 수 있다.

        - 회귀선의 기울기는 독립 변수의 변화에 따른 종속 변수의 변화를 나타낸다.

3) 오차항과 잔차의 정의

    - 오차항(𝜖): 실제 관측값과 모델이 예측한 값 간의 차이를 나타내며, 모델이 설명하지 못하는 부분이다.
    
        - 이는 우리가 측정할 수 없으며 모델이 가지는 불확실성을 나타낸다.

    - 잔차(Residual): 오차항의 추정치로, 실제 관측값과 모델 예측값의 차이이다. (=각 관측값에 대한 오차)

        - 잔차 $= Y - \hat{Y}$ (여기서 $\hat{Y}$는 회귀모형이 예측한 값이다.)

    - 요약: 오차항은 실제 값과 참된 회귀식 간의 차이이고, 잔차는 실제 값과 추정된 회귀식 간의 차이.

### 3. 회귀모형의 추정

1) 최소제곱법 (OLS: Ordinary Least Squares)

    - 최소제곱법은 회귀선이 데이터에 가장 잘 맞도록 하는 방법이다.
    
    - 데이터 포인트와 회귀선 사이의 거리(잔차)의 제곱합을 최소화하는 방향으로 회귀계수를 추정한다.
    
    - $\min \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2$

        - 여기서 $Y_i$는 실제 값, $\hat{Y_i}$는 예측 값이다.

2) 모형의 계수 추정

    - 최소제곱법을 통해 회귀계수(기울기와 절편)를 추정한다.
    
    - 이 값들은 독립 변수의 변화가 종속 변수에 미치는 영향을 설명한다.

    - OLS를 통해 구한 계수는 다음과 같이 계산된다.

        - 단순회귀의 경우: $\beta_1 = \frac{n(\sum XY) - (\sum X)(\sum Y)}{n(\sum X^2) - (\sum X)^2}$
        
        - $\beta_0$는 다음과 같이 계산됩니다
            
        - $\beta_0 = \bar{Y} - \beta_1 \bar{X}$

3) 계수의 해석

    - 계수는 독립 변수가 종속 변수에 미치는 영향을 나타낸다.
    
    - 예를 들어, $\beta_1$이 2라면, 독립 변수 $X$가 1 증가할 때 종속 변수 $Y$는 2 증가함을 의미한다.

        - 절편(𝛽0​): 독립 변수 X가 0일 때, Y의 예상값이다.

        - 기울기(𝛽1): X가 1단위 증가할 때 Y의 변화량이다.

> **중요**: 회귀모형이 유효하기 위해서는 몇 가지 가정이 필요하다.

### 4. 회귀모형의 가정

1) 선형성 가정: 독립 변수와 종속 변수 간의 관계는 선형적이어야 한다.

2) 독립성 가정: 오차항이 서로 독립적이어야 한다.

    - 즉, 하나의 오차가 다른 오차에 영향을 미치지 않아야 한다.

3) 등분산성 가정: 오차항의 분산이 일정해야 한다.
    
    - 즉, 모든 독립 변수 값에서 오차의 크기가 일정해야 한다.

4) 정상성 가정: 오차항이 정규분포를 따라야 한다.

    - 이를 통해 회귀계수의 유의성 검정을 정확히 수행할 수 있다.

5) 다중공선성 문제: 독립 변수들 간의 상관관계까 너무 높지 않아야 한다.

    - 다중공선성이 있으면 회귀계수의 추정이 불안정해진다.

### 5. 회귀모형의 진단 및 검증

1. 잔차 분석

    - 잔차 분석은 회귀모형의 적합성을 평가하는 방법이다.
    
    - 잔차의 패턴을 시각적으로 분석하여 정규성을 띄고, 패턴 없이 분포되어 있는지 확인한다.
    
    - 패턴이 있으면 선형성 가정이 위배되었을 수 있다.

2. 이상치와 영향치 탐지

    - 이상치는 데이터의 일반적인 패턴과 현저히 다른 값이다.
    
    - 영향치는 회귀모형에 큰 영향을 미치는 데이터이다.
    
    - 이들을 탐지하기 위해 Cook's Distance와 같은 방법을 사용할 수 있다.

3. 다중공선성 진단 (VIF: Variance Inflation Factor)

    - VIF는 다중공선성을 진단하는 지표로, 특정 독립 변수가 다른 독립 변수와 얼마나 상관관계가 있는지를 나타낸다.
    
    - 일반적으로 VIF 값이 10을 초과하면 다중공선성이 존재한다고 판단하여 문제가 있는 것으로 본다.

4. 모형의 적합도 평가 (R², Adjusted R²)

    - R²: 독립 변수들이 종속 변수를 얼마나 설명하는지 나타내며, 값이 1에 가까울수록 모형의 적합도가 높다.

    - Adjusted R²: 변수의 개수를 고려해 R²를 조정한 값으로, 변수 추가에 따른 과적합을 방지한다.

### 6. 회귀모형의 통계적 검정

1) F-검정 (`모형 전체`의 유의성 검정)

    - F-검정은 회귀모형이 통계적으로 유의미한지를 검정하는 방법이다.

    - 귀무가설은 "모든 회귀계수는 0이다"이다.

        - "모든 회귀계수는 0이다"라는 표현은 회귀모형에서 독립 변수가 종속 변수에 영향을 미치지 않는다는 의미이다.
        
        - 해당 문장을 자세하게 설명하자면 다음과 같다.

            1. 회귀모형의 기본 개념
            
                - 회귀모형은 독립 변수가 종속 변수에 미치는 영향을 분석하는 도구이다.
            
                - 회귀계수는 이러한 영향을 수치적으로 나타내는 값이다.

            2. 회귀계수의 의미
            
                - 각 독립 변수의 회귀계수는 그 변수가 종속 변수에 얼마나 영향을 미치는지를 나타낸다.
            
                - 예를 들어, 회귀계수가 2라면, 독립 변수가 1만큼 증가할 때 종속 변수는 2만큼 증가한다는 의미이다.

            3. 회귀계수가 0이라는 것의 의미

                - 모든 회귀계수가 0이라는 것은 모든 독립 변수가 종속 변수에 영향을 미치지 않는다는 뜻이다.
                
                - 즉, 독립 변수를 아무리 변화시켜도 종속 변수는 변화하지 않는다.

                - 이는 회귀모형이 의미가 없거나, 독립 변수들이 종속 변수와 관계가 없다는 것을 나타낸다.

            4. 결과 해석
            
                - 만약 회귀 분석의 결과로 모든 회귀계수가 0으로 나타난다면, 이는 모델이 잘못되었거나, 사용된 변수들이 적절하지 않다는 신호일 수 있다.
                
                - 이 경우 추가적인 변수를 고려하거나, 데이터의 질을 검토해야 할 필요가 있다.
    
    - 즉, 귀무가설은 "모형이 유의미하지 않다"와 같으며, 이는 회귀모형의 유효성을 의심하게 만드는 결과이다.

2) t-검정 (`회귀계수`의 유의성 검정)

    - 각 회귀계수가 통계적으로 유의한지 t-검정을 통해 확인한다.
    
    - 귀무가설은 "해당 회귀계수는 0이다"이다.

    - 각 회귀계수가 0인지 아닌지를 검정한다.
    
    - t-검정의 결과가 유의미하면 해당 독립 변수가 종속 변수에 영향을 미친다고 판단할 수 있다.

3) 신뢰구간과 예측구간

    - 신뢰구간: 회귀계수의 추정값에 대한 불확실성을 나타낸다.
    
    - 예측구간: 새로운 관측값에 대한 예측값의 범위를 나타냅니다.

### 7. 회귀모형의 확장

1) 다항 회귀 (Polynomial Regression)
    
    - 다항 회귀는 독립 변수의 다항식 형태로 회귀모형을 구성한다.

    - 즉, 선형이 아닌 `곡선 관계`를 표현한다.
    
    - 예를 들어, 2차 다항 회귀는 다음과 같이 표현된다.

    - $Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \epsilon$
    
2) 상호작용 효과 분석

    - 상호작용 효과는 두 개 이상의 독립 변수가 함께 영향을 미치는 경우를 분석한다.
    
    - 이를 위해 `두 독립 변수를 곱한 항`을 추가하여 회귀모형을 구성한다.
    
    - 예를 들어, 𝑌=𝛽0+𝛽1𝑋1+𝛽2𝑋2+𝛽3(𝑋1×𝑋2) 이다.

3) 범주형 변수와 더미 변수

    - 범주형 변수를 회귀모형에 포함시키기 위해 더미 변수를 사용한다.
    
    - 예를 들어, 성별을 표현할 때 '남성 = 0', '여성 = 1'로 코딩한다.

### 8. 모형 선택과 정규화

1) 변수 선택 방법

    - 모델에서 중요한 변수를 선택하는 방법이다. 전진 선택법, 후진 제거법, 단계별 선택법 등이 있다.

        1. 전진 선택법(Forward Selection)
        
            - 가장 유의미한 변수를 하나씩 추가하여 최적의 모형을 구축한다.
            
            - 첫 번째 단계에서는 모든 독립 변수를 비교하여 가장 유의미한 변수를 선택하고,
            
            - 그 다음 단계에서는 선택된 변수에 대해 추가적인 변수를 고려한다.

        2. 후진 제거법(Backward Elimination)
        
            - 모든 변수를 포함하여 시작한 후, 가장 유의미하지 않은 변수를 차례로 제거한다.
            
            - 이 방법은 모델의 복잡성을 줄이면서도 충분한 설명력을 유지하는 데 유용하다.

        3. 단계별 선택법(Stepwise Selection)
        
            - 전진 선택법과 후진 제거법을 결합한 방법으로, 변수를 추가하거나 제거하는 과정에서 유의미성을 평가한다.
            
            - 이 방법은 최적의 변수 조합을 찾는 데 도움을 준다.

2) 과적합 문제와 교차 검증

    - 과적합(Overfitting): 모델이 훈련 데이터에 너무 잘 맞춰져 새로운 데이터에 대한 예측력이 떨어지는 문제다.
    
        - 이를 방지하기 위해 교차 검증(Cross-Validation) 기법을 사용할 수 있다.
        
        - 데이터셋을 여러 부분으로 나누고, 각 부분을 검증 데이터로 사용하여 모델의 일반화 성능을 평가한다.

3) 릿지 회귀와 라소 회귀

    - 릿지 회귀: 회귀계수의 크기를 제어하여 과적합을 방지한다.

        - L2 정규화를 사용하여 회귀 계수의 크기를 제한함으로써 과적합을 방지한다.
        
        - 비용 함수는 다음과 같이 표현된다.

            - $\text{Cost Function} = \sum (Y_i - \hat{Y_i})^2 + \lambda \sum \beta_j^2$

            - 여기서 $\lambda$는 정규화 강도를 조절하는 하이퍼파라미터이다.

    - 라소 회귀: 불필요한 변수의 회귀계수를 0으로 만들어 자동으로 변수 선택을 수행한다.

        - L1 정규화를 사용하여 계수를 0으로 만들 수 있는 특성이 있다.
        
        - 이로 인해 불필요한 변수를 자동으로 제거하는 기능이 있다.
        
        - 비용 함수는 다음과 같다.

            - $\text{Cost Function} = \sum (Y_i - \hat{Y_i})^2 + \lambda \sum |\beta_j|$

### 9. 비선형 회귀분석

1) 비선형 회귀모형의 개념

    - 비선형 회귀모형은 독립 변수와 종속 변수 간의 관계가 선형이 아닐 때 사용된다.
    
    - 예를 들어, 지수 함수나 로그 함수로 표현되는 경우이다.
    
    - 다음과 같은 형태의 비선형 회귀모형이 있다.

        - $Y = \beta_0 + \beta_1 e^{\beta_2 X} + \epsilon$

2) 비선형 회귀모형의 추정 방법

    - 비선형 회귀모형의 계수 추정은 최소제곱법 대신 일반적으로 비선형 최적화 기법을 사용하여 수행된다.
    
    - 예를 들어, 뉴턴-랩슨 방법이나 경사 하강법을 활용하여 최소화 문제를 해결한다.

3) 비선형 모형의 응용 사례

    - 비선형 회귀모형은 생물학에서 성장 곡선을 분석하거나 공학에서 기계의 수명을 예측, 경제적 수요 함수, 머신 러닝의 다양한 모델에서 사용됩니다.
    
    - 예를 들어, 로지스틱 회귀는 이진 분류 문제에 적합한 비선형 회귀모형이다.

### 10. 회귀분석의 실제 적용

1) 데이터 전처리와 변환

    - 회귀분석을 수행하기 전에 데이터 전처리가 필요하다.
    
    - 결측값 처리, 변수 변환(예: 로그 변환), 스케일링(정규화 또는 표준화), 이상치 제거 등이 포함된다.
    
    - 이러한 과정은 모델의 성능을 높이고 가정의 충족을 보장한다.

2) 회귀분석 실습 (Python/Numpy 예제)

    - 다음은 NumPy를 활용한 간단한 단순 회귀분석 예제이다.
    
    - 이 예제에서는 가상의 데이터를 생성하고, 최소제곱법을 통해 회귀선을 추정한다.

```python
# 환경 설정
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 가상의 데이터 생성
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # 독립 변수
y = 4 + 3 * X + np.random.randn(100, 1)  # 종속 변수

# 선형 회귀 모델을 위한 계수 추정 (최소제곱법)
X_b = np.c_[np.ones((100, 1)), X]  # X에 절편 추가
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

# 예측값 계산
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # 절편 추가
y_predict = X_new_b.dot(theta_best)

# 결과 시각화
plt.plot(X_new, y_predict, "r-", label="예측선")
plt.plot(X, y, "b.", label="데이터 포인트")
plt.xlabel("독립 변수")
plt.ylabel("종속 변수")
plt.legend()
plt.title("단순 회귀분석")
plt.show()
```

```python
# 환경 설정
import numpy as np
from numpy.linalg import inv

# 가상 데이터 생성 (단순 회귀 예제)
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([[1.2], [2.4], [3.1], [3.9], [5.1]])

# 행렬 형태로 변환
X_b = np.c_[np.ones((5, 1)), X]  # 절편을 포함한 X 행렬

# 최소제곱법으로 회귀계수 계산
theta_best = inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

print(f"회귀계수 (절편과 기울기): \n{theta_best}")
```

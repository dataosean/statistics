### 1. ÌôïÎ•† Î∂ÑÌè¨Ïùò Í∞úÏöî

    - Îç∞Ïù¥ÌÑ∞Ïùò Ìå®ÌÑ¥ÏùÑ Ïù¥Ìï¥ÌïòÍ≥† ÏòàÏ∏°Ìï† Ïàò ÏûàÍ≤å Ìï¥Ï£ºÎ©∞, Îç∞Ïù¥ÌÑ∞ Í≥ºÌïôÏóêÏÑú ÌÜµÍ≥ÑÏ†Å Î™®Îç∏ÎßÅÍ≥º Î®∏Ïã†Îü¨ÎãùÏóê Ï§ëÏöîÌïú Ïó≠Ìï†ÏùÑ ÌïúÎã§.

### 2. Ïù¥ÏÇ∞ ÌôïÎ•† Î∂ÑÌè¨

1) Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨(Bernoulli Distribution)

    - Îëê Í∞ÄÏßÄ Í∞ÄÎä•Ìïú Í≤∞Í≥º(Ïòà: ÏÑ±Í≥µ/Ïã§Ìå®)Î•º Í∞ÄÏßÑ ÏÇ¨Í±¥Ïùò ÌôïÎ•†ÏùÑ Î™®Îç∏ÎßÅÌïúÎã§.
    
    - ÏòàÎ•º Îì§Ïñ¥, ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ÏóêÏÑú ÏïûÎ©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•†Ïù¥ 0.5ÎùºÎ©¥, Ïù¥Îäî Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§.

```python
from scipy.stats import bernoulli
p = 0.5  # ÏÑ±Í≥µ ÌôïÎ•†
data = bernoulli.rvs(p, size=100)  # 100Î≤àÏùò Ïã§Ìóò
data
```

2) Ïù¥Ìï≠ Î∂ÑÌè¨ (Binomial Distribution)

    - Ïù¥Ìï≠ Î∂ÑÌè¨Îäî Ïó¨Îü¨ Î≤àÏùò ÎèÖÎ¶ΩÏ†ÅÏù∏ Î≤†Î•¥ÎàÑÏù¥ Ïã§ÌóòÏùò `ÏÑ±Í≥µ ÌöüÏàò`Î•º Î™®Îç∏ÎßÅÌïúÎã§.
    
    - ÏòàÎ•º Îì§Ïñ¥, 10Î≤àÏùò ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ÏóêÏÑú 5Î≤à ÏïûÎ©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞Ìï† Ïàò ÏûàÎã§.

```python
from scipy.stats import binom
n = 10  # Ïã§Ìóò ÌöüÏàò
p = 0.5  # ÏÑ±Í≥µ ÌôïÎ•†
data = binom.rvs(n, p, size=100)  # 100Î≤àÏùò Ïã§Ìóò
data
```

3) Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ (Poisson Distribution)

    - Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Îäî ÌäπÏ†ï ÏãúÍ∞Ñ ÎÇ¥Ïóê Î∞úÏÉùÌïòÎäî ÏÇ¨Í±¥Ïùò ÌöüÏàòÎ•º Î™®Îç∏ÎßÅÌïúÎã§.
    
    - ÏòàÎ•º Îì§Ïñ¥, Ìïú ÏãúÍ∞Ñ ÎèôÏïàÏóê Í≥†Í∞ùÏù¥ Îß§Ïû•Ïóê Îì§Ïñ¥Ïò§Îäî ÌöüÏàòÎ•º ÏòàÏ∏°Ìï† Ïàò ÏûàÎã§.

```python
from scipy.stats import poisson
mu = 3  # ÌèâÍ∑† ÏÇ¨Í±¥ Î∞úÏÉù ÌöüÏàò
data = poisson.rvs(mu, size=100)
data
```

### 3. Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨

1) Ï†ïÍ∑ú Î∂ÑÌè¨ (Normal Distribution)

    - Í∞ÄÏû• ÎßéÏù¥ ÏÇ¨Ïö©ÎêòÎäî Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨ Ï§ë ÌïòÎÇòÏù¥Î©∞, ÌèâÍ∑†ÏùÑ Ï§ëÏã¨ÏúºÎ°ú ÎåÄÏπ≠Ï†ÅÏù∏ Ï¢Ö Î™®ÏñëÏùÑ ÎùàÎã§.
    
    - ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®Î°ú Ï†ïÏùòÎêòÎ©∞, ÎßéÏùÄ ÏûêÏó∞ ÌòÑÏÉÅÏù¥ Ï†ïÍ∑ú Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§.

```python
from scipy.stats import norm
mu, sigma = 0, 1  # ÌèâÍ∑†Í≥º ÌëúÏ§ÄÌé∏Ï∞®
data = norm.rvs(mu, sigma, size=100)
data
```

2) ÏßÄÏàò Î∂ÑÌè¨ (Exponential Distribution)

    - ÏßÄÏàò Î∂ÑÌè¨Îäî ÏÇ¨Í±¥ Í∞ÑÏùò ÏãúÍ∞Ñ Í∞ÑÍ≤©ÏùÑ Î™®Îç∏ÎßÅÌïòÎ©∞, Ï£ºÎ°ú ÎåÄÍ∏∞ ÏãúÍ∞ÑÏù¥ÎÇò ÏàòÎ™ÖÏù¥ ÏûàÎäî Îç∞Ïù¥ÌÑ∞Ïóê ÏÇ¨Ïö©ÎêúÎã§.

    - ÏöîÏïΩ: ÏÇ¨Í±¥Ïù¥ Î∞úÏÉùÌï† ÎïåÍπåÏßÄÏùò ÎåÄÍ∏∞ ÏãúÍ∞Ñ Î∂ÑÌè¨.

```python
from scipy.stats import expon
scale = 1  # ÌèâÍ∑†
data = expon.rvs(scale=scale, size=100)
data
```

3) Í∑†Îì± Î∂ÑÌè¨ (Uniform Distribution)

    - Î™®Îì† Í∞íÏù¥ ÎèôÏùºÌïú(Í∑†Îì±Ìïú) ÌôïÎ•†Î°ú Î∞úÏÉùÌï† ÌôïÎ•†ÏùÑ Í∞ÄÏßÄÎäî Î∂ÑÌè¨Î°ú, ÌäπÏ†ï Íµ¨Í∞ÑÏóêÏÑú ÎûúÎç§Ìïú Í∞íÏùÑ ÏÉùÏÑ±ÌïòÎäî Îç∞ Ïú†Ïö©ÌïòÎã§.

```python
from scipy.stats import uniform
low, high = 0, 1  # Î≤îÏúÑ
data = uniform.rvs(low, high, size=100)
data
```

### 4. ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò(PDF)ÏôÄ ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF)

1) PDF(Probability Density Function) - ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò

    - PDFÎäî ÌäπÏ†ï Í∞í Í∑ºÏ≤òÏóêÏÑú ÌôïÎ•† Î≥ÄÏàòÍ∞Ä Î∞úÏÉùÌï† Í∞ÄÎä•ÏÑ±ÏùÑ ÎÇòÌÉÄÎÇ∏Îã§.
    
    - Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨ÏóêÏÑú ÏÇ¨Ïö©ÎêòÎ©∞, ÌäπÏ†ï Í∞íÏù¥ Î∞úÏÉùÌï† ÌôïÎ•†ÏùÄ 0Ïù¥ÎØÄÎ°ú Íµ¨Í∞ÑÏùò ÌôïÎ•†ÏùÑ Í≥ÑÏÇ∞ÌïúÎã§.

2) CDF(Cumulative Distribution Function) - ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò

    - CDFÎäî ÌäπÏ†ï Í∞í Ïù¥ÌïòÏùò Í∞íÏù¥ ÎÇòÏò¨ ÌôïÎ•†ÏùÑ ÎÇòÌÉÄÎÇ¥Î©∞, ÌôïÎ•† Î∂ÑÌè¨Î•º ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïù¥Ìï¥ÌïòÎäî Îç∞ ÎèÑÏõÄÏùÑ Ï§ÄÎã§.

3) Ïù¥ÏÇ∞ Î∂ÑÌè¨Îäî ÌôïÎ•† ÏßàÎüâ Ìï®Ïàò (PMF)Î°ú ÌëúÌòÑ, Ïó∞ÏÜç Î∂ÑÌè¨Îäî ÌôïÎ•† Î∞ÄÎèÑ Ìï®Ïàò (PDF)Î°ú ÌëúÌòÑÌïúÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.linspace(-3, 3, 100)
pdf = norm.pdf(data, 0, 1)
cdf = norm.cdf(data, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(data, pdf, label='PDF')
plt.plot(data, cdf, label='CDF')
plt.legend()
plt.title('PDF and CDF of Normal Distribution')  # Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò PDFÏôÄ CDF
plt.show()
```

### 5. Í∏∞ÎåìÍ∞í, Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®

1) Í∏∞ÎåìÍ∞í (Mean)

    - ÌôïÎ•† Î∂ÑÌè¨Ïùò Ï§ëÏã¨ Í≤ΩÌñ•ÏÑ±ÏùÑ ÎÇòÌÉÄÎÇ¥Î©∞, Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Î•º ÌèâÍ∑†Ï†ÅÏúºÎ°ú ÎåÄÌëúÌïòÎäî Í∞íÏù¥Îã§.

2) Î∂ÑÏÇ∞ (Variance)

    - Îç∞Ïù¥ÌÑ∞Í∞Ä Í∏∞ÎåìÍ∞íÏóêÏÑú ÏñºÎßàÎÇò Îñ®Ïñ¥Ï†∏ ÏûàÎäîÏßÄÎ•º ÎÇòÌÉÄÎÇ¥Îäî Í∞íÏúºÎ°ú, Î∂ÑÌè¨Ïùò ÌçºÏßê Ï†ïÎèÑÎ•º Ï∏°Ï†ïÌïúÎã§.

3) ÌëúÏ§ÄÌé∏Ï∞® (Standard Deviation)

    - Î∂ÑÏÇ∞Ïùò Ï†úÍ≥±Í∑ºÏúºÎ°ú, Îç∞Ïù¥ÌÑ∞Ïùò Î≥ÄÎèôÏÑ±ÏùÑ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïù¥Ìï¥ÌïòÍ∏∞ ÏâΩÍ≤å Ìï¥Ï§ÄÎã§.

```python
import numpy as np

data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}, Variance: {variance}, Standard Deviation: {std_dev}")
```

### 6. Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ (Central Limit Theorem)

1) Ï∂©Î∂ÑÌûà ÌÅ∞ ÏÉòÌîå ÌÅ¨Í∏∞Î•º Í∞ÄÏßÑÎã§Î©¥, ÏÉòÌîåÏùò ÌèâÍ∑† Î∂ÑÌè¨Í∞Ä ÏõêÎûò Î∂ÑÌè¨ÏôÄ ÏÉÅÍ¥ÄÏóÜÏù¥ Ï†ïÍ∑ú Î∂ÑÌè¨Ïóê Í∞ÄÍπåÏõåÏßÑÎã§Îäî Ïù¥Î°†Ïù¥Îã§.
    
2) Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑùÏóêÏÑú ÏÉòÌîåÎßÅÏùÑ ÌÜµÌï¥ ÌÜµÍ≥ÑÏ†Å Ï∂îÎ°†ÏùÑ Ìï† Ïàò ÏûàÎäî Í∑ºÍ±∞Í∞Ä ÎêúÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt

means = [np.mean(np.random.exponential(1, 1000)) for _ in range(1000)]
plt.hist(means, bins=30, edgecolor='k')
plt.title('Distribution of Sample Means')
plt.show()
```

### 7. Î≤†Ïù¥ÏßÄÏïà ÌôïÎ•†Í≥º Ï°∞Í±¥Î∂Ä ÌôïÎ•†

1) Î≤†Ïù¥Ï¶à Ï†ïÎ¶¨

    - Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Î∞îÌÉïÏúºÎ°ú ÏÇ¨Í±¥Ïùò ÌôïÎ•†ÏùÑ `Í∞±Ïã†`ÌïòÎäî Î∞©Î≤ïÏùÑ Ï†úÍ≥µÌïúÎã§.
    
    - Îç∞Ïù¥ÌÑ∞ Í≥ºÌïôÏóêÏÑúÎäî ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º Î∞òÏòÅÌï¥ Î™®Îç∏ÏùÑ ÏóÖÎç∞Ïù¥Ìä∏Ìï† Îïå Ïú†Ïö©ÌïòÎã§.

    - Î®∏Ïã†Îü¨Îãù ÌïôÏäµ 10ÌöåÏ∞® ÍπÉÌóàÎ∏å ÏóÖÎ°úÎìú Î¨∏ÏÑú - ÎÇòÏù¥Î∏å Î≤†Ïù¥Ï¶à Î™®Îç∏ Ï∞∏Í≥†ÌïòÍ∏∞.

### 10. ÏòàÏ†ú ÏΩîÎìú(1)

1) ÏïÑÎûò ÏΩîÎìúÏóêÏÑúÎäî Ï†ïÍ∑ú Î∂ÑÌè¨, ÏßÄÏàò Î∂ÑÌè¨, Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±ÌïòÍ≥†,

2) Í∞ÅÍ∞ÅÏùò Î∂ÑÌè¨Î•º ÌûàÏä§ÌÜ†Í∑∏Îû®Í≥º Ïª§ÎÑê Î∞ÄÎèÑ Ï∂îÏ†ïÏùÑ ÌÜµÌï¥ ÏãúÍ∞ÅÌôîÌïúÎã§.

3) Ìï¥Îãπ ÏòàÏ†úÎäî Îã§ÏñëÌïú ÌôïÎ•† Î∂ÑÌè¨Ïùò ÌäπÏÑ±ÏùÑ ÏãúÍ∞ÅÏ†ÅÏúºÎ°ú ÎπÑÍµêÌïòÍ≥† Ïù¥Ìï¥ÌïòÎäî Îç∞ ÎèÑÏõÄÏùÑ Ï§ÄÎã§.

```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ï†ïÍ∑ú Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞
norm_data = np.random.normal(0, 1, 1000)
# ÏßÄÏàò Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞
exp_data = np.random.exponential(1, 1000)
# Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞
poisson_data = np.random.poisson(3, 1000)

# Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(15, 5))

# Ï†ïÍ∑ú Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
plt.subplot(1, 3, 1)
sns.histplot(norm_data, kde=True)
plt.title('Normal Distribution')
# ÏßÄÏàò Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
plt.subplot(1, 3, 2)
sns.histplot(exp_data, kde=True)
plt.title('Exponential Distribution')
# Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ ÏãúÍ∞ÅÌôî
plt.subplot(1, 3, 3)
sns.histplot(poisson_data, kde=True)
plt.title('Poisson Distribution')

plt.tight_layout()
plt.show()
```

```python
# ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy.stats import norm

# Îç∞Ïù¥ÌÑ∞ÏÖã Î∂àÎü¨Ïò§Í∏∞
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# 'petal length' Ïª¨Îüº ÏÑ†ÌÉù
petal_length = df['petal length (cm)']

# Îç∞Ïù¥ÌÑ∞Ïùò Í∏∞Ï¥à ÌÜµÍ≥Ñ ÌôïÏù∏ÌïòÍ∏∞
mean = petal_length.mean()
std = petal_length.std()

print(f"Mean: {mean}")
print(f"Standard Deviation: {std}")
```

```python
# ÌûàÏä§ÌÜ†Í∑∏Îû®Í≥º Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥°ÏÑ† Í∑∏Î¶¨Í∏∞
sns.histplot(petal_length, kde=False, stat='density', bins=20, color='blue')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)

title = f"Fit results: mean = {mean:.2f},  std = {std:.2f}"
plt.title(title)
plt.xlabel('Petal Length(cm)')
plt.ylabel('Density')
plt.show()
```

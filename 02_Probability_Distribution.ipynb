### 1. 확률 분포의 개요

1) 확률 분포란?

    - 어떤 확률 변수가 취할 수 있는 모든 가능한 값과 각각의 값이 발생할 확률을 설명하는 함수.
    
    - 데이터의 패턴을 이해하고 예측할 수 있게 해주며, 데이터 과학에서 통계적 모델링과 머신러닝에 중요한 역할을 한다.

2) 데이터 분석에서의 중요성

    - 확률 분포를 이해하면 데이터의 특성을 파악하고, 이상치 탐지, 가설 검정, 예측 모델링 등 다양한 데이터 분석 작업에서 정확도를 높일 수 있다.

### 2. 이산 확률 분포

1) 베르누이 분포(Bernoulli Distribution)

    - 두 가지 가능한 결과(예: 성공/실패)를 가진 사건의 확률을 모델링한다.
    
    - 예를 들어, 동전 던지기에서 앞면이 나올 확률이 0.5라면, 이는 베르누이 분포를 따른다.

```python
from scipy.stats import bernoulli
p = 0.5  # 성공 확률
data = bernoulli.rvs(p, size=100)  # 100번의 실험
data
```

2) 이항 분포 (Binomial Distribution)

    - 이항 분포는 여러 번의 독립적인 베르누이 실험의 `성공 횟수`를 모델링한다.
    
    - 예를 들어, 10번의 동전 던지기에서 5번 앞면이 나올 확률을 계산할 수 있다.

```python
from scipy.stats import binom
n = 10  # 실험 횟수
p = 0.5  # 성공 확률
data = binom.rvs(n, p, size=100)  # 100번의 실험
data
```

3) 포아송 분포 (Poisson Distribution)

    - 포아송 분포는 특정 시간 내에 발생하는 사건의 횟수를 모델링한다.
    
    - 예를 들어, 한 시간 동안에 고객이 매장에 들어오는 횟수를 예측할 수 있다.

```python
from scipy.stats import poisson
mu = 3  # 평균 사건 발생 횟수
data = poisson.rvs(mu, size=100)
data
```

### 3. 연속 확률 분포

1) 정규 분포 (Normal Distribution)

    - 가장 많이 사용되는 연속 확률 분포 중 하나이며, 평균을 중심으로 대칭적인 종 모양을 띈다.
    
    - 평균과 표준편차로 정의되며, 많은 자연 현상이 정규 분포를 따른다.

```python
from scipy.stats import norm
mu, sigma = 0, 1  # 평균과 표준편차
data = norm.rvs(mu, sigma, size=100)
data
```

2) 지수 분포 (Exponential Distribution)

    - 지수 분포는 사건 간의 시간 간격을 모델링하며, 주로 대기 시간이나 수명이 있는 데이터에 사용된다.

    - 요약: 사건이 발생할 때까지의 대기 시간 분포.

```python
from scipy.stats import expon
scale = 1  # 평균
data = expon.rvs(scale=scale, size=100)
data
```

3) 균등 분포 (Uniform Distribution)

    - 모든 값이 동일한(균등한) 확률로 발생할 확률을 가지는 분포로, 특정 구간에서 랜덤한 값을 생성하는 데 유용하다.

```python
from scipy.stats import uniform
low, high = 0, 1  # 범위
data = uniform.rvs(low, high, size=100)
data
```

### 4. 확률 밀도 함수(PDF)와 누적 분포 함수(CDF)

1) PDF(Probability Density Function) - 확률 밀도 함수

    - PDF는 특정 값 근처에서 확률 변수가 발생할 가능성을 나타낸다.
    
    - 연속 확률 분포에서 사용되며, 특정 값이 발생할 확률은 0이므로 구간의 확률을 계산한다.

2) CDF(Cumulative Distribution Function) - 누적 분포 함수

    - CDF는 특정 값 이하의 값이 나올 확률을 나타내며, 확률 분포를 직관적으로 이해하는 데 도움을 준다.

3) 이산 분포는 확률 질량 함수 (PMF)로 표현, 연속 분포는 확률 밀도 함수 (PDF)로 표현한다.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.linspace(-3, 3, 100)
pdf = norm.pdf(data, 0, 1)
cdf = norm.cdf(data, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(data, pdf, label='PDF')
plt.plot(data, cdf, label='CDF')
plt.legend()
plt.title('PDF and CDF of Normal Distribution')  # 정규분포의 PDF와 CDF
plt.show()
```

### 5. 기댓값, 분산, 표준편차

1) 기댓값 (Mean)

    - 확률 분포의 중심 경향성을 나타내며, 전체 데이터를 평균적으로 대표하는 값이다.

2) 분산 (Variance)

    - 데이터가 기댓값에서 얼마나 떨어져 있는지를 나타내는 값으로, 분포의 퍼짐 정도를 측정한다.

3) 표준편차 (Standard Deviation)

    - 분산의 제곱근으로, 데이터의 변동성을 직관적으로 이해하기 쉽게 해준다.

```python
import numpy as np

data = np.random.normal(0, 1, 1000)
mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print(f"Mean: {mean}, Variance: {variance}, Standard Deviation: {std_dev}")
```

### 6. 중심극한정리 (Central Limit Theorem)

1) 충분히 큰 샘플 크기를 가진다면, 샘플의 평균 분포가 원래 분포와 상관없이 정규 분포에 가까워진다는 이론이다.
    
2) 데이터 분석에서 샘플링을 통해 통계적 추론을 할 수 있는 근거가 된다.

```python
import numpy as np
import matplotlib.pyplot as plt

means = [np.mean(np.random.exponential(1, 1000)) for _ in range(1000)]
plt.hist(means, bins=30, edgecolor='k')
plt.title('Distribution of Sample Means')
plt.show()
```

### 7. 베이지안 확률과 조건부 확률

1) 베이즈 정리

    - 주어진 데이터를 바탕으로 사건의 확률을 `갱신`하는 방법을 제공한다.
    
    - 데이터 과학에서는 새로운 데이터를 반영해 모델을 업데이트할 때 유용하다.

    - 머신러닝 학습 10회차 깃허브 업로드 문서 - 나이브 베이즈 모델 참고하기.

2) 조건부 확률

    - 한 사건이 발생했을 때 다른 사건이 발생할 확률로, 데이터 간의 관계를 분석하는 데 사용된다.

```python
from scipy.stats import bayes_mvs

data = np.random.normal(0, 1, 1000)
mean, _, _ = bayes_mvs(data, alpha=0.95)
print(f"Mean: {mean}")
```

### 8. 정규성 가정과 정규성 검정

1) 정규성의 중요성
    
    - 많은 통계적 방법이 데이터가 정규 분포를 따른다는 가정을 한다.
    
    - 따라서 데이터를 분석하기 전 정규성을 확인하는 것이 중요하다.

2) 정규성 검정 방법

    - 샤피로-윌크 검정과 Q-Q 플롯을 사용해 데이터가 정규 분포를 따르는지 확인할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(0, 1, 1000)
stats.probplot(data, dist="norm", plot=plt)
plt.show()
```

### 9. 확률 분포의 시각화

1) 히스토그램

    - 데이터를 구간별로 나누어 빈도를 시각화하는 데 사용되며, 데이터의 분포를 직관적으로 보여준다.

```python
import seaborn as sns

data = np.random.normal(0, 1, 1000)
sns.histplot(data, kde=True)
plt.title('Histogram and Kernel Density Estimation')
plt.show()
```

### 10. 예제 코드(1)

1) 아래 코드에서는 정규 분포, 지수 분포, 포아송 분포 데이터를 생성하고,

2) 각각의 분포를 히스토그램과 커널 밀도 추정을 통해 시각화한다.

3) 해당 예제는 다양한 확률 분포의 특성을 시각적으로 비교하고 이해하는 데 도움을 준다.

```python
# 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 정규 분포 데이터
norm_data = np.random.normal(0, 1, 1000)
# 지수 분포 데이터
exp_data = np.random.exponential(1, 1000)
# 포아송 분포 데이터
poisson_data = np.random.poisson(3, 1000)

# 데이터 시각화
plt.figure(figsize=(15, 5))

# 정규 분포 시각화
plt.subplot(1, 3, 1)
sns.histplot(norm_data, kde=True)
plt.title('Normal Distribution')
# 지수 분포 시각화
plt.subplot(1, 3, 2)
sns.histplot(exp_data, kde=True)
plt.title('Exponential Distribution')
# 포아송 분포 시각화
plt.subplot(1, 3, 3)
sns.histplot(poisson_data, kde=True)
plt.title('Poisson Distribution')

plt.tight_layout()
plt.show()
```

### 11. 예제 코드(2)

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy.stats import norm

# 데이터셋 불러오기
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# 'petal length' 컬럼 선택
petal_length = df['petal length (cm)']

# 데이터의 기초 통계 확인하기
mean = petal_length.mean()
std = petal_length.std()

print(f"Mean: {mean}")
print(f"Standard Deviation: {std}")
```

```python
# 히스토그램과 정규분포 곡선 그리기
sns.histplot(petal_length, kde=False, stat='density', bins=20, color='blue')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)

title = f"Fit results: mean = {mean:.2f},  std = {std:.2f}"
plt.title(title)
plt.xlabel('Petal Length(cm)')
plt.ylabel('Density')
plt.show()
```

1) 히스토그램을 그린 후, 정규분포 곡선을 히스토그램 위에 그려서 데이터가 정규분포를 따르는지 시각적으로 확인했다.

2) 히스토그램과 정규분포 곡선을 비교해 보면, petal length의 값이 정규분포와 유사하게 분포하고 있음을 확인할 수 있다.

3) 그러나 데이터의 분포는 완벽한 정규분포는 아니며, 특히 극단 값이나 비대칭성도 보인다.

### 💡`게임 이론(Game Theory)` 요약

1. 게임 이론은 여러 플레이어가 각기 다른 전략을 선택할 때의 상호작용을 분석하는 수학적 이론이다.

2. 주된 개념으로는 플레이어, 전략, 보수가 있으며, 내쉬 균형은 모든 플레이어가 최적의 전략을 선택하는 상황을 설명한다.

4. 베이즈 게임은 불확실한 정보에서의 의사 결정을, 진화 게임 이론은 생물학적 진화를 모델링한다.

5. 게임 이론은 경제학, 정치학, 생물학 등 다양한 분야에 적용된다.

### 1. 게임 이론의 개요

1) 게임 이론의 정의와 역사

    - 게임 이론은 플레이어들이 전략을 선택할 때의 결과를 분석하는 학문이다.
    
    - 경제학자 존 폰 노이만과 오스카 모르겐슈테른이 1944년에 출판한 "게임 이론과 경제 행동"에서 체계적으로 정립되었다.

    - 원래 군사 전략 및 경제적 결정의 분석을 위해 개발되었다. 이후 정치학, 생물학 등 다른 분야로도 확장되었다.

2) 게임 이론의 주요 개념 (플레이어, 전략, 보수)

    - 게임 이론에서 중요한 요소는 다음과 같다.
    
        - 플레이어: 게임에 참여하는 모든 개체(인간, 기업, 국가 등)
        
        - 전략: 각 플레이어가 선택할 수 있는 모든 가능한 행동

        - 보수: 각 전략 조합에 따라 플레이어가 얻는 이익이나 손해. 보수는 각 플레이어의 선택과 상호작용에 의해 결정된다.
        
    - 예를 들어, 두 회사가 시장에서 경쟁할 때, 각 회사는 가격을 어떻게 설정할지(전략) 결정하고, 그 결과로 얻는 이익(보수)을 계산한다.

3) 게임 이론의 응용 분야

    - 게임 이론은 경제학, 정치학, 생물학, 컴퓨터 과학 등 다양한 분야에서 중요한 역할을 한다.

        - 경제학: 기업의 가격 설정, 독점적 경쟁 분석.

        - 정치학: 선거 전략, 국제 협상.

        - 생물학: 생물 간의 경쟁과 협력, 진화적 안정 전략.

        - 컴퓨터 과학: 알고리즘의 효율성 분석, 네트워크 안정성.

    - 이 외에도 게임 이론은 사회적 선택 이론, 재난 관리, 교육 정책 등 다양한 분야에서도 응용된다.

### 2. 전략적 게임의 기본 원리

1) 순수 전략과 혼합 전략

    - 순수 전략은 플레이어가 특정 행동을 확실하게 선택하는 경우이다.
    
    - 반면, 혼합 전략은 여러 행동을 일정한 확률로 선택하는 경우를 말한다.
    
    - 예를 들어, 가위바위보 게임에서 혼합 전략은 가위, 바위, 보를 각각 1/3의 확률로 선택하는 것이다.

2) 내쉬 균형(Nash Equilibrium)의 정의와 중요성

    - 내쉬 균형은 각 플레이어가 자신의 전략을 바꿀 동기가 없는 상태를 의미한다.
    
    - 즉, 모든 플레이어가 자신의 전략을 고수하는 것이 최선의 선택일 때 내쉬 균형이 성립한다.
    
    - 이 개념은 많은 게임에서 중요한 분석 도구로 사용된다.

3) 지배 전략(Strict Dominance)과 약한 지배 전략(Weak Dominance)

    - 지배 전략(Strict Dominance)
    
        - 특정 전략이 다른 모든 전략보다 항상 더 나은 결과를 가져오는 경우, 그 전략을 지배 전략이라고 한다.
    
        - 즉, 어떤 상황에서도 이 전략을 선택하는 것이 가장 유리하다.

        - 강한 지배 전략(Strictly Dominant Strategy)과 동일한 개념으로, 전략이 다른 모든 전략보다 확실히 우수한 경우를 말한다.

        - 예를 들어, 광고를 하는 것이 광고를 하지 않는 것보다 항상 더 높은 수익을 가져온다면, 광고를 하는 것이 지배 전략이 된다.

    - 약한 지배 전략(Weak Dominance)
    
        - 특정 전략이 다른 전략과 같은 결과를 가져오거나 더 나은 결과를 가져오는 경우, 그 전략을 약한 지배 전략이라고 한다.

### 3. 협력적 게임과 비협력적 게임

1) 협력적 게임의 정의와 특성

    - 협력적 게임에서는 플레이어들이 협력하여 공동의 이익을 극대화하려고 한다.
    
    - 협력적 게임의 주요 요소는 협력과 연합을 통한 공동의 목표 달성이다.

    - 보수 분배: 협력적 게임에서는 플레이어들 간에 얻은 이익을 어떻게 분배할지가 중요한 문제이다.
    
        - 이를 위해 여러 가지 방법이 있으며, 코어(Core), 샤플리 값(Shapley Value) 등이 사용된다.

    - 계약과 연합: 플레이어들이 계약을 맺거나 연합을 형성하여 공동의 목표를 달성하고 이익을 공유한다.
    
        - 연합 형성은 게임의 성격에 따라 다양한 전략적 결정이 필요하다.

2) 비협력적 게임의 정의와 특성

    - 비협력적 게임에서는 각 플레이어가 독립적으로 행동하며, 협상이나 연합이 불가능하다.
    
    - 이 경우, 각 플레이어는 자신의 전략을 최적화하려고 한다.

    - 독립적 행동: 플레이어는 자신의 선택이 다른 플레이어의 선택에 영향을 받지 않고, 자신의 보수를 극대화하기 위해 최선의 전략을 선택한다.

    - 내쉬 균형: 비협력적 게임에서는 내쉬 균형이 중요한 개념이다.
    
        - 내쉬 균형은 모든 플레이어가 자신의 전략을 고수할 때, 어떤 플레이어도 자신의 전략을 변경할 유인이 없는 상태를 의미한다.

3) 협상 이론(Negotiation Theory)과 균형 해석

    - 협상 이론은 협력적 게임에서 발생할 수 있는 다양한 상황을 분석한다.
    
    - 협상 과정에서 어떤 전략이 효과적일지, 협상 결과는 어떻게 나타날지를 분석하는 데 사용된다.

    - 협상 과정: 협상 이론은 플레이어들이 어떻게 협상하고, 합의에 도달하는 과정을 분석한다.
    
        - 협상에서의 전략적 결정, 상호작용 및 합의 도출 방법 등을 연구한다.

    - 균형 해석: 협상 이론에서의 균형 해석은 협상 결과가 어떻게 형성되는지,
    
        - 그리고 각 플레이어가 협상 과정에서 어떤 전략을 채택해야 하는지를 이해하는 데 도움을 준다.

### 4. 반복 게임과 동적 게임

1) 반복 게임(Repeated Games)의 개념과 중요성

    - 반복 게임은 동일한 게임이 여러 번 반복되는 상황을 말한다.
    
    - 반복 게임에서는 한 번의 게임 결과가 이후 게임의 전략에 영향을 미칠 수 있다.
    
    - 이를 통해 플레이어들은 장기적인 이익을 고려하여 전략을 조정할 수 있다.

        - 신뢰와 협력: 반복 게임에서는 플레이어들이 서로 신뢰를 구축하고 협력할 수 있는 기회를 가질 수 있다.
        
            - 예를 들어, "죄수의 딜레마" 같은 게임에서는 반복 게임에서 상호 협력이 이루어질 수 있다.

        - 보복과 협상: 반복 게임에서는 한 플레이어가 다른 플레이어의 행동에 대해 보복하거나 협상할 수 있는 전략을 채택할 수 있다.
        
            - 이는 협력적 행동을 유도할 수 있다.

2) 동적 게임(Dynamic Games)에서의 정보 구조 (완전 정보와 불완전 정보)

    - 동적 게임에서는 게임이 여러 단계로 진행되며, 각 단계에서 플레이어는 이전의 행동을 고려하여 전략을 선택한다.
    
    - 정보 구조는 완전 정보와 불완전 정보로 나눌 수 있다.

        - 완전 정보 게임: 모든 플레이어가 게임의 모든 상황에 대한 정보를 알고 있는 경우이다.
        
            - 예를 들어, 체스와 같은 게임이 이에 해당한다.

        - 불완전 정보 게임: 일부 정보만 공개되어 있거나 플레이어들이 게임의 상태를 완전히 알지 못하는 경우이다.
        
            - 이 경우, 플레이어는 불확실성을 감안하여 전략을 선택해야 한다.

3) 하향 귀납법(Backward Induction)과 동적 게임의 균형

    - 하향 귀납법: 동적 게임에서 마지막 단계부터 거꾸로 올라가면서 최적의 전략을 찾는 방법이다.
    
        - 이 방법을 통해 각 단계에서 최적의 전략을 선택할 수 있다.
        
        - 예를 들어, 체스 게임에서 마지막 단계부터 플레이어가 최적의 수를 찾는 방식이 이에 해당한다.

    - 동적 게임의 균형: 하향 귀납법을 사용하여 찾은 전략은 동적 게임의 균형을 의미한다.
        
        - 이는 게임의 모든 단계에서 플레이어들이 최적의 전략을 취했을 때, 전체 게임의 균형 상태를 설명한다.

### 5. 베이즈 게임(Bayesian Games)

1) 불확실성 하에서의 게임 이론

    - 베이즈 게임은 플레이어들이 불확실한 정보에 근거해 전략을 선택하는 상황을 다룬다.
    
        - 예를 들어, 각 플레이어가 다른 플레이어의 타입(성향)을 알지 못할 때 베이즈 게임이 적용된다.

    - 플레이어의 타입: 각 플레이어는 자신만 알고 있는 정보를 포함하여 자신의 타입(예: 성향, 능력)과 관련된 정보를 가지고 있으며, 이 정보가 게임의 전략 선택에 영향을 미친다.

    - 정보 비대칭: 모든 플레이어가 동일한 정보를 갖지 않고, 각 플레이어가 가진 정보의 차이가 전략 선택에 영향을 미치게 된다.

2) 베이즈 균형(Bayesian Equilibrium)과 그 해석

    - 베이즈 균형은 각 플레이어가 불확실한 정보 하에서도 최적의 전략을 선택하는 균형 상태를 의미한다.
    
    - 이는 내쉬 균형의 확장 개념으로, 각 플레이어는 다른 플레이어의 행동에 대한 확률적 가정을 바탕으로 최적의 행동을 선택한다.

        - 확률적 가정: 각 플레이어는 다른 플레이어의 타입이 확률 분포로 주어질 때, 이 확률 분포에 기반하여 최적의 전략을 선택한다.

        - 균형 상태: 모든 플레이어가 자신의 전략을 최적화하면, 다른 플레이어들이 자신의 정보와 전략에 기반하여 최적화된 전략을 선택하는 상태가 베이즈 균형이다.

3) 신호 게임(Signaling Games)과 정보 비대칭

    - 신호 게임은 정보 비대칭이 존재하는 상황에서 한 플레이어가 신호를 보내고, 다른 플레이어가 그 신호를 해석하여 전략을 결정하는 게임이다.
    
    - 신호 게임은 주로 경제학에서 사용되며, 예를 들어 구직자가 자신의 능력을 신호로 전달하는 상황을 모델링한다.
    
        - 신호와 신호 수신자: 신호를 보내는 플레이어(신호 송신자)는 자신의 타입이나 능력에 대한 정보를 신호로 제공하며, 신호를 받는 플레이어(신호 수신자)는 이 신호를 바탕으로 결정을 내린다.

        - 정보 비대칭 해소: 신호 게임은 정보 비대칭 문제를 해결하기 위한 방법을 제시하며, 신호의 정확성이나 신뢰성에 따라 균형이 달라질 수 있다.

### 6. 진화 게임 이론

1) 진화 안정 전략(Evolutionarily Stable Strategy, ESS)의 개념

    - 진화 안정 전략(ESS)은 집단 내에서 대다수 개체가 이 전략을 사용할 때, 다른 전략이 도입되어도 성공적으로 번식하지 못하거나 생존하지 못하는 전략을 의미한다.
    
    - 즉, ESS는 집단 내에서 안정적인 상태를 유지할 수 있는 전략이다.

    - 이는 생물학적 진화와 관련된 상황에서 많이 사용된다.

        - 진화 안정성: ESS는 단순히 최적의 전략이 아니라, 다른 전략이 도입되었을 때에도 여전히 안정성을 유지할 수 있는 전략이다.

        - 생물학적 예시: 예를 들어, 특정 동물의 공격과 방어 전략이 ESS가 될 수 있다.
        
            - 공격 전략을 사용하는 대다수의 동물들은 이러한 전략을 통해 생존 가능성이 높아지며,
            
            - 다른 방어 전략이 도입되더라도 경쟁에서 우위를 점할 수 있다.

2) 진화 게임 이론의 통계적 해석

    - 진화 게임 이론에서는 특정 전략이 시간이 지남에 따라 어떻게 확산되는지를 분석한다.
    
    - 이때 통계적 방법을 사용하여 전략의 성공 확률, 생존율 등을 모델링할 수 있다.

    - 예를 들어, 특정 유전자형이 어떻게 번식할지 등을 모델링할 수 있다.

        - 모델링: 유전자형의 빈도 변화, 전략의 전파 등 장기적인 변화를 분석하기 위해 확률론적 모델링, 시뮬레이션 기법 등이 사용된다.

        - 통계적 기법: 예를 들어, 몬테 카를로 시뮬레이션을 사용하여 다양한 전략의 장기적인 성공률을 예측할 수 있다.

3) 게임 이론에서의 학습과 적응

    - 플레이어들이 반복되는 게임 과정에서 전략을 학습하고 적응하는 방법을 분석한다.
    
    - 학습 모델을 통해 시간에 따라 전략이 어떻게 변화하고 최적화되는지를 이해할 수 있다.

        - 학습 모델: 강화 학습, 적응적 전략 모델 등의 기법을 사용하여 플레이어들이 환경에 따라 어떻게 전략을 조정하는지를 분석한다.

        - 적응 과정: 반복 게임에서는 플레이어들이 상대방의 전략을 학습하고, 이에 대응하여 자신의 전략을 조정해 나가는 과정을 분석한다.

### 7. 게임 이론의 수학적 도구

1) 유틸리티 이론(Utility Theory)과 기대 유틸리티

    - 유틸리티 이론: 플레이어가 다양한 선택지에서 얻는 만족도(유틸리티)를 수치화하여 분석하는 방법이다.
    
        - 유틸리티는 플레이어의 선호도를 반영하며, 각 선택의 가치나 만족도를 표현한다.

    - 기대 유틸리티 이론: 불확실한 상황에서 각 선택지가 가져올 만족도의 기대값을 계산하고, 이를 기반으로 최적의 전략을 선택한다.
    
        - 확률적 요소가 포함되어 각 선택의 평균 유틸리티를 계산한다.

        - 기대 유틸리티는 다음과 같이 계산된다.

        - 

        - 𝐸𝑈는 기대 유틸리티, 𝑃(𝑖)는 결과 𝑖의 발생 확률, 𝑈(𝑖)는 결과 𝑖의 유틸리티이다.

    - 유틸리티 함수: 유틸리티 이론에서는 유틸리티 함수를 사용하여 각 선택지의 유틸리티를 정의한다.
    
        - 유틸리티 함수는 플레이어의 선호를 수치화하는 함수이다.

2) 매트릭스 게임의 수학적 표현

    - 매트릭스 게임은 두 명의 플레이어가 각각 선택할 수 있는 전략과 그 결과를 행렬 형태로 나타낸 것이다.

    - 각 셀은 특정 전략 조합에 대한 보수를 나타내며, 이를 통해 각 플레이어의 최적 전략을 쉽게 분석할 수 있다.

    - 매트릭스 게임은 일반적으로 두 개의 행렬로 표현된다.
    
        - 하나는 첫 번째 플레이어의 보수를, 다른 하나는 두 번째 플레이어의 보수를 나타낸다.
        
        - 예를 들어, 2×2 매트릭스 게임에서는 다음과 같은 형태로 나타낼 수 있다.

        - 

        - (𝑎𝑖𝑗,𝑏𝑖𝑗)는 전략 𝑖와 𝑗의 조합에 따른 두 플레이어의 보수이다.

3) 미니맥스 정리(Minimax Theorem)와 혼합 전략 균형

    - 미니맥스 정리: 제로섬 게임에서 각 플레이어가 자신의 손실을 최소화하려는 전략을 설명하는 정리이다.
    
        - 이 정리는 특정 전략이 최적의 방어 전략을 제공함을 보장한다.

        - 제로섬 게임에서는 플레이어가 상대방의 최악의 경우를 최소화하려고 한다.
        
        - 미니맥스 정리는 이 균형을 설명하며, 수학적으로 다음과 같은 식으로 표현된다.

        - 

        - 𝑓(𝑥,𝑦)는 전략 𝑥와 𝑦의 조합에 대한 보수이다.

    - 혼합 전략 균형: 플레이어가 여러 전략을 혼합하여 사용하는 상황에서의 균형 상태를 설명한다.
    
        - 각 플레이어는 자신의 전략 혼합 비율을 조정하여 상대방의 최적 전략에 대응한다.

            - 혼합 전략 균형은 각 플레이어가 여러 전략을 일정한 확률로 사용하여 전략을 혼합하는 상황을 다룬다.
        
            - 플레이어들은 자신의 혼합 전략을 조정하여 상대방의 혼합 전략에 맞춰 균형을 이루려고 한다.

### 8. 게임 이론의 통계적 응용

1) 경매 이론(Auction Theory)과 최적 입찰 전략

    - 경매 이론: 다양한 경매 형식에서 플레이어들이 최적의 입찰 전략을 찾는 방법을 분석한다.
    
        - 경매 이론은 입찰자의 전략이 경매의 결과에 미치는 영향을 이해하고 최적의 입찰가를 결정하는 데 도움을 준다.

    - 경매 유형: 경매 이론에서는 다양한 경매 유형이 분석된다.
    
        - 예를 들어, 첫 가격 경매와 두 번째 가격 경매(비다윈 경매)가 있다.
        
        - 첫 가격 경매에서는 가장 높은 입찰가를 제시한 사람이 경매를 승리하지만,
        
        - 두 번째 가격 경매에서는 두 번째로 높은 입찰가를 제시한 사람이 최종 낙찰가가 된다.

    - 최적 입찰 전략: 경매에서 최적의 입찰 전략을 결정하는 방법에는 정보의 비대칭을 고려하는 것이 포함된다.
    
        - 예를 들어, 경매 참여자가 자신의 입찰이 다른 참여자들의 입찰에 어떻게 영향을 미칠지를 분석한다.

2) 경쟁 시장 분석에서의 게임 이론 적용

    - 경쟁 시장 분석: 기업들이 경쟁하는 시장에서 가격, 제품, 마케팅 전략 등을 어떻게 설정하는지 분석한다.
    
        - 게임 이론은 경쟁 시장에서 기업들이 전략을 어떻게 조정하는지를 설명하는 데 유용하다.

    - 가격 설정: 기업들이 가격을 설정할 때, 경쟁자의 가격 결정 전략에 따라 자신의 가격 전략을 조정하는 방법을 분석한다.
    
        - 가격 담합이나 가격 경쟁과 같은 전략적 상호작용을 이해하는 데 사용된다.

    - 게임 이론의 적용: 내쉬 균형을 통해 기업들이 가격이나 제품 전략을 조정할 때 발생할 수 있는 결과를 예측할 수 있다.
    
        - 예를 들어, 쿠폰 정책이나 제품 차별화 전략 등이 분석된다.

3) 사회적 선택 이론(Social Choice Theory)과 투표 제도 분석

    - 사회적 선택 이론: 여러 사람의 선호를 집합적으로 어떻게 반영할지 다루며, 사회적 의사 결정을 내리는 과정에서 게임 이론을 활용한다.

    - 투표 제도 분석: 다양한 투표 방식이 어떤 결과를 낳을지를 분석한다.
    
        - 예를 들어, 단순 다수제(plurality voting)와 가장 선호하는 방식(ranked choice voting) 등에서 결과가 어떻게 달라지는지를 연구한다.

    - 투표 제도의 특성: 투표 제도는 각기 다른 방식으로 사회의 선호를 반영한다.
    
        - Borda Count 또는 Condorcet 방법 같은 투표 방식은 다양한 방법으로 후보자들의 선호를 집합적으로 반영한다.

    - 파레토 효율성: 사회적 선택 이론에서는 파레토 효율성을 기준으로 결정을 내리는 방법을 분석한다.
    
        - 이는 어떤 개인의 후생을 증가시키면서 다른 개인의 후생을 감소시키지 않는 상황을 의미한다.

### 9. 게임 이론과 데이터 분석

1) 빅데이터에서의 전략적 의사 결정

    - 데이터 기반 의사 결정: 빅데이터를 활용한 전략적 의사 결정에서 게임 이론은 경쟁적 상호작용과 협력을 모델링하는 데 유용하다.
    
        - 예를 들어, 고객 세그먼트에 대한 데이터를 분석하여 각 세그먼트에 맞는 맞춤형 마케팅 전략을 개발할 수 있다.

    - 경쟁 분석: 데이터 분석을 통해 경쟁사의 전략을 예측하고, 이를 기반으로 자신의 전략을 조정하는 것도 게임 이론의 적용 사례이다.

2) 머신 러닝과 게임 이론의 통합

    - 게임 이론과 머신 러닝을 통합하여 복잡한 의사 결정 문제를 해결하는 방법을 연구한다.

        - 복잡한 의사 결정: 게임 이론은 여러 에이전트 간의 상호작용을 분석할 수 있으며, 머신 러닝은 데이터에서 패턴을 학습하여 예측과 의사 결정을 개선할 수 있다. 이 두 가지 접근법의 통합은 복잡한 의사 결정 문제를 해결하는 데 효과적이다.
    
    - 예를 들어, 자율주행차의 경로 결정 문제에서 게임 이론과 강화 학습을 결합할 수 있다.

        - 강화 학습과 게임 이론: 자율주행차에서의 경로 결정 문제를 해결할 때, 게임 이론은 다양한 상황에서의 전략적 상호작용을 모델링하고, 강화 학습은 이 모델을 학습하여 최적의 경로를 결정하는 데 사용될 수 있다.

3) 게임 이론을 활용한 최적화 문제 해결

    - 게임 이론을 사용하여 다양한 최적화 문제를 해결한다.

        - 전략적 상호작용: 게임 이론은 전략적 상호작용을 모델링하고, 이를 통해 최적의 결정을 내리는 방법을 분석한다.
        
            - 이는 데이터 분석과 결합되어 실제 문제 해결에 활용될 수 있다.
    
    - 예를 들어, 자원 배분 문제에서 각 플레이어가 최적의 결정을 내리도록 모델링할 수 있다.

        - 자원 배분: 게임 이론을 사용하여 자원 배분 문제를 모델링할 때, 각 플레이어의 목표와 제약 조건을 고려하여 최적의 자원 배분 전략을 찾을 수 있다.
        
            - 예를 들어, 경매나 가격 책정 문제에서도 게임 이론을 활용하여 최적의 자원 배분을 결정할 수 있다.

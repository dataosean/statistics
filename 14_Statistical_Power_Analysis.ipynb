### 5. Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù ÎèÑÍµ¨ Î∞è ÏÜåÌîÑÌä∏Ïõ®Ïñ¥

1) R, Python Îì± ÌîÑÎ°úÍ∑∏ÎûòÎ∞ç Ïñ∏Ïñ¥ÏóêÏÑúÏùò Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù

    - PythonÏùò statsmodels Ìå®ÌÇ§ÏßÄÎ•º ÏÇ¨Ïö©Ìï¥ Í≤ÄÏ†ïÎ†•ÏùÑ Î∂ÑÏÑùÌï† Ïàò ÏûàÎã§.
    
    - ÏòàÎ•º Îì§Ïñ¥, t-Í≤ÄÏ†ïÏùò Í≤ÄÏ†ïÎ†•ÏùÑ Í≥ÑÏÇ∞ÌïòÎäî ÏΩîÎìúÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

        - effect_size = 0.5Îäî Îëê ÏßëÎã® Í∞Ñ ÌèâÍ∑† Ï∞®Ïù¥Í∞Ä 0.5 ÌëúÏ§ÄÌé∏Ï∞®ÎßåÌÅº ÎÇúÎã§Îäî ÏùòÎØ∏Ïù¥Îã§.

        - alpha = 0.05Îäî 1Ï¢Ö Ïò§Î•òÎ•º 5% Ïù¥ÌïòÎ°ú ÏÑ§Ï†ïÌïú Í∞íÏù¥Îã§.

        - power = 0.8ÏùÄ Ïó∞Íµ¨ÏûêÍ∞Ä 80%Ïùò Í≤ÄÏ†ïÎ†•ÏùÑ Î™©ÌëúÎ°ú ÌïúÎã§Îäî ÏùòÎØ∏Ïù¥Îã§.

```python
# pythonÏùò statsmedols Ìå®ÌÇ§ÏßÄ Î∂àÎü¨Ïò§Í∏∞
import numpy as np
from statsmodels.stats.power import TTestIndPower

# ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞
effect_size = 0.5  # Ìö®Í≥º ÌÅ¨Í∏∞
alpha = 0.05       # Ïú†ÏùòÏàòÏ§Ä
power = 0.8        # Î™©Ìëú Í≤ÄÏ†ïÎ†•

# ÌïÑÏöîÌïú ÌëúÎ≥∏ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power)
print(f'ÌïÑÏöîÌïú ÌëúÎ≥∏ ÌÅ¨Í∏∞: {sample_size}')
```

        3. G*Power: ÏßÅÍ¥ÄÏ†Å Ïù∏ÌÑ∞ÌéòÏù¥Ïä§, Îπ†Î•∏ Í≥ÑÏÇ∞ Í∞ÄÎä•

4) ÌÜµÍ≥ÑÏ†Å Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù Í≤∞Í≥ºÏùò ÏãúÍ∞ÅÌôî

    - Í≤ÄÏ†ïÎ†• Í≥°ÏÑ† Îì±ÏùÑ ÏãúÍ∞ÅÌôîÌïòÏó¨ Ïó∞Íµ¨ÏûêÏôÄ Ïù¥Ìï¥Í¥ÄÍ≥ÑÏûêÏóêÍ≤å Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù Í≤∞Í≥ºÎ•º Î™ÖÌôïÌïòÍ≤å Ï†ÑÎã¨Ìï† Ïàò ÏûàÎã§.

    - ÏòàÎ•º Îì§Ïñ¥, matplotlibÏôÄ scipy.statsÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Í≤ÄÏ†ïÎ†• Í≥°ÏÑ†ÏùÑ Í∑∏Î¶¥ Ïàò ÏûàÎã§.

        1. ÏÑ§Ï†ï: Ïú†ÏùòÏàòÏ§Ä(Œ±)Í≥º Ìö®Í≥º ÌÅ¨Í∏∞(Ïòà: Cohen's d)Î•º ÏÑ§Ï†ïÌïúÎã§.

        2. ÏÉòÌîå ÌÅ¨Í∏∞: ÏÉòÌîå ÌÅ¨Í∏∞ Î≤îÏúÑÎ•º Ï†ïÏùòÌïúÎã§.

        3. Í≤ÄÏ†ïÎ†• Í≥ÑÏÇ∞: Í∞Å ÏÉòÌîå ÌÅ¨Í∏∞Ïóê ÎåÄÌï¥ Í∑ÄÎ¨¥Í∞ÄÏÑ§Í≥º ÎåÄÎ¶ΩÍ∞ÄÏÑ§Ïùò Î∂ÑÌè¨Î•º Ï†ïÏùòÌïòÍ≥†, Í≤ÄÏ†ïÎ†•ÏùÑ Í≥ÑÏÇ∞ÌïúÎã§.

        4. ÏãúÍ∞ÅÌôî: matplotlibÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Í≤ÄÏ†ïÎ†• Í≥°ÏÑ†ÏùÑ Í∑∏Î¶¨Í≥†, Í≤ÄÏ†ïÎ†•Ïù¥ 0.8 Ïù¥ÏÉÅÏù∏ ÏÉòÌîå ÌÅ¨Í∏∞Î•º Í∞ïÏ°∞ÌïúÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import koreanize_matplotlib

# ÏÑ§Ï†ï
alpha = 0.05  # Ïú†ÏùòÏàòÏ§Ä
effect_size = 0.5  # Ìö®Í≥º ÌÅ¨Í∏∞ (Ïòà: Cohen's d)
n = np.arange(10, 200, 5)  # ÏÉòÌîå ÌÅ¨Í∏∞ Î≤îÏúÑ
power = []

# Í≤ÄÏ†ïÎ†• Í≥ÑÏÇ∞
for sample_size in n:
    # Í∑ÄÎ¨¥Í∞ÄÏÑ§ Î∂ÑÌè¨
    null_mean = 0
    null_std = 1 / np.sqrt(sample_size)

    # ÎåÄÎ¶ΩÍ∞ÄÏÑ§ Î∂ÑÌè¨
    alternative_mean = effect_size
    alternative_std = 1 / np.sqrt(sample_size)

    # Í≤ÄÏ†ïÎ†• Í≥ÑÏÇ∞: ÎåÄÎ¶ΩÍ∞ÄÏÑ§ ÌïòÏóêÏÑúÏùò Í∏∞Í∞ÅÏó≠
    critical_value = stats.norm.ppf(1 - alpha, loc=null_mean, scale=null_std)
    power_value = stats.norm.cdf(critical_value, loc=alternative_mean, scale=alternative_std)
    power.append(1 - power_value)

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.plot(n, power, label='Í≤ÄÏ†ïÎ†• Í≥°ÏÑ†', color='blue')
plt.title('Í≤ÄÏ†ïÎ†• Î∂ÑÏÑù Í≤∞Í≥º')
plt.xlabel('ÏÉòÌîå ÌÅ¨Í∏∞ (n)')
plt.ylabel('Í≤ÄÏ†ïÎ†• (Power)')
plt.axhline(y=0.8, color='red', linestyle='--', label='Í≤ÄÏ†ïÎ†• 0.8 Í∏∞Ï§ÄÏÑ†')
plt.axvline(x=n[np.argmax(np.array(power) >= 0.8)], color='green', linestyle='--',
            label='Í≤ÄÏ†ïÎ†• 0.8Ïóê ÎèÑÎã¨ÌïòÎäî ÏÉòÌîå ÌÅ¨Í∏∞')
plt.legend()
plt.grid()
plt.show()
```

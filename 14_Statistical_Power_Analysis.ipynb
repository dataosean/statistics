###  1. 통계적 검정력의 개요

1) 통계적 검정력의 정의

    - 통계적 검정력(Power)은 `귀무가설이 거짓일 때 이를 올바르게 기각할 확률`을 의미한다.
    
        - 즉, 실제로 효과가 존재하는 경우, 연구에서 그 효과를 발견할 수 있는 능력을 나타낸다.

    - 이는 `가설 검정에서 실제로 존재하는 효과를 발견할 수 있는 능력`과도 같은 의미이다.
    
        - 검정력이 높을수록 실제 효과를 발견할 확률이 높아진다.
    
        - 검정력은 0과 1 사이의 값을 가지며, 1에 가까울수록 좋다.
    
    - 검정력(Power) = 1 - 2종 오류(β)로 정의된다.

        - 즉, 검정력이 높다는 것은 2종 오류를 줄일 수 있다는 의미이다.

2) 통계적 검정력의 중요성

    - 통계적 검정력은 연구 설계의 품질을 판단하는 매우 중요한 요소이다.
    
    - 낮은 검정력은 실제로 존재하는 효과를 발견하지 못할 가능성을 높이며, 이는 연구 결과의 신뢰도를 떨어뜨린다.

3) 1종 오류, 2종 오류 개념 및 검정력의 관계

    - 1종 오류(α): 귀무가설이 참일 때, 이를 잘못 기각하는 오류.

    - 2종 오류(β): 귀무가설이 거짓임에도, 이를 기각하지 못하는 오류.

    - 검정력(Power): 2종 오류를 범하지 않을 확률로, 1 - β로 계산된다.

    

4) 통계적 검정력과 유의수준(α)의 관계

    - 요약: 유의수준(α)이 낮아질수록 1종 오류의 확률이 줄어들지만, 반대로 2종 오류의 확률은 증가할 수 있다. 따라서, α를 조정하면 검정력에도 영향을 미치게 되니 주의하여 설정한다.

    - 유의수준(α)과 검정력은 서로 상관관계를 가진다.

        - `유의수준`이란?
        
            - 예를 들어, α가 0.05라면, 귀무가설이 실제로 참일 때 5%의 확률로 기각할 수 있다는 의미이다.
            
            - 이것은 귀무가설이 실제로 참이라면, 이론적으로 100번의 실험 중 95번은 귀무가설을 유지하고, 5번은 우연히 기각될 수 있음을 의미한다.

            - 즉, 일부 경우에는 우연에 의해 통계적 검정에서 귀무가설을 기각할 수 있다는 말이다.

            - 통계적 검정은 항상 불확실성을 동반한다. 실제로 귀무가설이 참이라도, 데이터의 변동성이나 샘플링 오류로 인해 잘못된 결정을 내릴 수 있다. 이 때문에 유의수준을 설정하여 이러한 오류의 가능성을 미리 정의하는 것이다.

        - `통계적 검정`이란?
            
            - 연구자는 데이터를 수집한 후, 이 데이터를 바탕으로 귀무가설이 참인지 아닌지를 판단한다. 이때, 통계적 검정을 통해 p-값을 계산한다.
    
    - α를 낮추면 1종 오류의 가능성을 줄일 수 있지만, 그만큼 검정력도 감소하게 된다.

        - WHY? 유의수준(α)을 낮추면 1종 오류의 가능성이 줄어드는 이유는, 유의수준(α)이 1종 오류를 허용하는 최대 확률이기 때문이다.

            - 즉, α를 낮추면 귀무가설을 기각하기 위한 `기준이 더 엄격`해져서, 실제로 귀무가설이 참인 경우에 기각하는 오류의 가능성이 줄어든다.

            - α를 낮추면 귀무가설을 기각하기 위한 기준이 더 엄격해진다는 의미는 다음과 같다.

                1. 기각 기준의 강화: α는 귀무가설이 참일 때 이를 기각할 최대 확률을 나타내며, 일반적으로 0.05, 0.01 등의 값을 사용한다. α를 0.01로 설정하면, 귀무가설을 기각하기 위해서 더 강력한 결과를 요구하게 된다.

                2. 통계적 검정의 엄격함: α를 낮추면, 통계적 검정에서 효과가 있다고 판단하기 위해 필요한 p-값(p-value)이 더 작아져야 한다. 따라서, 같은 데이터라도 효과를 발견하기 어려워진다.

            - 결과적으로, α를 낮추면 귀무가설을 기각하기 위해 요구되는 증거의 수준이 높아져, 통계적 검정이 더 엄격해진다는 의미이다.

        - WHY? 하지만 α를 낮추면 검정력(1 - β, 즉 2종 오류의 가능성을 줄이는 능력)이 감소하는 이유는 아래와 같다.

            1. 기준 강화: α를 낮추면 통계 검정에서 효과가 있다고 판단하기 위해 더 많은 증거가 필요해진다. 즉, 귀무가설을 기각하기 위해 요구되는 기준이 높아진다.

            2. 실제 효과 발견의 어려움: 귀무가설이 거짓일 때, 즉 실제로 효과가 존재할 때도 α가 낮아지면 더 엄격한 기준 때문에 그 효과를 발견하기 어려워진다. 그래서 실제로 효과가 있음에도 불구하고 귀무가설을 채택하게 되는 경우(2종 오류의 가능성)가 많아지게 된다.

        - 결과적으로, α를 낮추면 1종 오류는 줄어들지만, 반대로 실제 효과가 있을 때 이를 놓칠 가능성(즉, 2종 오류)이 증가하게 된다. 이렇게 되면 검정력이 감소하는 것이다. 두 가지 오류 간의 균형을 맞추는 것이 통계 분석에서 중요한 부분이다.

            1. 1종 오류가 줄어든다: 귀무가설이 참일 때, 제대로 채택한다. = 귀무가설 채택.

            2. 2종 오류가 줄어든다: 귀무가설이 거짓일 때, 올바르게 기각한다. = 대립가설 채택.
    
    - 연구 설계 시 적절한 유의수준(α)과 검정력을 균형 있게 설정하는 것이 중요하다.

### ⭐tip.극단적인 결과의 확률과 유의 수준 비교 총정리

1. 개념

    1) 유의수준(α)

        - 귀무가설이 참일 때 기각할 수 있는 최대 확률.

        - 연구자가 설정하는 값으로, 일반적으로 0.05, 0.01 등이 사용된다.

    2) 극단적인 결과의 확률

        - 귀무가설이 참일 때, 실험 결과와 같거나 더 극단적인 값이 나타날 확률.
        
        - p-value라고 하며, 실제 실험 결과에서 계산된 값이다.

2. 예시

    1) 가정

        - 유의수준 α = 0.05
        
            - 귀무가설을 기각할 기준으로 5%의 확률을 설정한다.

        - 극단적인 결과의 확률 = 6%
        
            - 귀무가설이 참일 때, 우연히 극단적인 값이 나타날 확률.

            - 우연일 확률이 클수록 귀무가설이 참일 가능성이 높아진다고 해석할 수 있다.
            
            - 귀무가설이 참일 때, 실험 결과와 같거나 더 극단적인 결과가 나타날 확률.

            - 이 확률이 클수록 (가정으로 세운) 귀무가설이 참임을 뒷받침해준다.

3. 해석

    1) 유의수준과 비교

        - `유의수준 α = 0.05`는 기각할 수 있는 최대 허용 확률이다.


        - `극단적인 결과의 확률 = 6%`는 유의수준 α인 5%를 초과한다.


    2) 귀무가설 기각 여부

        - 유의수준 5%를 설정했으므로, (우연히) 극단적인 결과가 5% 이하의 확률로 발생할 경우에만 귀무가설을 기각할 수 있다.

        - 실험 결과, 극단적인 결과의 확률이 6%라면?
        
            - p값이 유의수준인 5%보다 크므로, 귀무가설을 기각하지 않는다.
            
            - 이는 현재의 결과가 귀무가설이 맞다는 가정 하에서 우연히 발생할 수 있는 결과로 간주되기 때문이다.

4. 결론

    - 극단적인 결과의 확률이 유의수준 α보다 크다면, 즉 6%의 확률로 극단적인 결과가 발생할 수 있다면, 이는 귀무가설을 기각할 만한 충분한 증거가 없다는 의미이다.

    - 유의수준 α = 0.05에서 6%의 확률은 귀무가설이 맞을 때 나타날 수 있는 결과의 범위 내에 있기 때문에, 귀무가설이 틀렸다고 결론짓기에는 부족한 증거이다.

    - 따라서, 5%의 유의수준을 설정했는데, 극단적인 결과가 6%의 확률로 발생한다면? 현재의 실험 결과로는 귀무가설을 기각하지 않는다.
    
    - 이는 결과가 귀무가설 하에서 우연히 발생할 가능성이 여전히 5%를 초과하기 때문이다.

### 2. 검정력 분석의 기본 개념

1) 효과 크기(Effect Size) 개념

    - 효과 크기는 연구에서 관찰된 현상의 크기 또는 중요도를 나타내는 지표이다.
    
    - 예를 들어, 두 그룹 간의 평균 차이, 상관계수 등이 이에 해당한다.
    
    - 효과 크기가 클수록 동일한 표본 크기에서 검정력이 증가한다.

2) 표본 크기(Sample Size)와 검정력

    - 표본 크기가 클수록 검정력이 증가한다.
    
    - 이는 표본 크기가 클수록 모집단의 특징을 더 정확하게 반영하기 때문이다.

3) 유의수준(Significance Level)과 검정력

    - 유의수준(α)이 낮아지면 1종 오류의 위험은 감소하지만, 그만큼 검정력도 함께 감소한다.
    
    - 검정력이 높을수록 2종 오류를 줄일 수 있기 때문이다.

    - 따라서 연구 설계 시 α와 검정력을 적절히 설정하는 것이 중요하다.

4) 검정력 곡선(Power Curve)

    - 검정력 곡선은 효과 크기, 표본 크기, 유의수준 간의 관계에 따른 검정력 변화를 시각적으로 표현한 것이다.

    - 일반적으로 표본 크기가 증가할수록 검정력도 증가하는 곡선 형태를 가진다.
    
    - 이 곡선을 통해 연구자는 원하는 검정력을 달성하기 위해 필요한 표본 크기 등을 확인할 수 있다.

2) 실용적 접근: 표본 크기 산정

    - 표본 크기 산정은 검정력 분석의 중요한 부분이다.
    
    - 이는 원하는 검정력 수준을 유지하면서 유의미한 결과를 도출하기 위해 필수적이다.

    - 표본 크기를 산정하기 위해서는 효과 크기, 유의수준, 검정력을 고려해야 한다.
    
    - 일반적으로 통계 소프트웨어를 사용하여 계산한다.

3) 표본 크기와 효과 크기의 상관관계

    - 효과 크기가 클수록 작은 표본 크기로도 충분한 검정력을 확보할 수 있다.
    
    - 반면, 효과 크기가 작으면 더 큰 표본이 필요하다.

4) 효과 크기 추정 방법 및 선택 기준

    - 효과 크기는 이전 연구 결과나 파일럿 연구, 도메인 지식을 바탕으로 추정한다.
    
    - Cohen의 d, Hedges' g, odds ratio 등이 효과 크기를 측정하는 데 사용된다.

    - 일반적으로 Cohen의 기준(작은 효과: 0.2, 중간 효과: 0.5, 큰 효과: 0.8)을 사용한다.

### 4. 검정력 분석의 절차

1) 가설 설정

    - 연구의 목적에 따라 명확한 귀무가설과 대립가설을 설정한다.

### 5. 검정력 분석 도구 및 소프트웨어

1) R, Python 등 프로그래밍 언어에서의 검정력 분석

    - Python의 statsmodels 패키지를 사용해 검정력을 분석할 수 있다.
    
    - 예를 들어, t-검정의 검정력을 계산하는 코드는 다음과 같다.

        - effect_size = 0.5는 두 집단 간 평균 차이가 0.5 표준편차만큼 난다는 의미이다.

        - alpha = 0.05는 1종 오류를 5% 이하로 설정한 값이다.

        - power = 0.8은 연구자가 80%의 검정력을 목표로 한다는 의미이다.

```python
# python의 statsmedols 패키지 불러오기
import numpy as np
from statsmodels.stats.power import TTestIndPower

# 예제 데이터
effect_size = 0.5  # 효과 크기
alpha = 0.05       # 유의수준
power = 0.8        # 목표 검정력

# 필요한 표본 크기 계산
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power)
print(f'필요한 표본 크기: {sample_size}')
```

```python
# python의 statsmedols 패키지 불러오기(위와 같음, 살짝 다른)
import numpy as np
import statsmodels.api as sm

# 효과 크기와 표본 크기 설정
effect_size = 0.5  # 효과 크기
alpha = 0.05       # 유의수준
power = 0.8        # 목표 검정력

# 필요한 표본 크기 계산
required_n = sm.stats.tt_ind_solve_power(effect_size=effect_size, alpha=alpha, power=power)
print(f"필요한 표본 크기: {required_n:.2f}")
```

2) G*Power, PASS 등 전문 소프트웨어 사용법

    - 프로그램마다 계산의 정확도, 사용 편의성, 기능의 다양성 등이 다르므로, 연구 목적에 맞는 도구를 선택해야 한다.

4) 통계적 검정력 분석 결과의 시각화

    - 검정력 곡선 등을 시각화하여 연구자와 이해관계자에게 검정력 분석 결과를 명확하게 전달할 수 있다.

    - 예를 들어, matplotlib와 scipy.stats를 사용하여 검정력 곡선을 그릴 수 있다.

        1. 설정: 유의수준(α)과 효과 크기(예: Cohen's d)를 설정한다.

        2. 샘플 크기: 샘플 크기 범위를 정의한다.

        3. 검정력 계산: 각 샘플 크기에 대해 귀무가설과 대립가설의 분포를 정의하고, 검정력을 계산한다.

        4. 시각화: matplotlib을 사용하여 검정력 곡선을 그리고, 검정력이 0.8 이상인 샘플 크기를 강조한다.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import koreanize_matplotlib

# 설정
alpha = 0.05  # 유의수준
effect_size = 0.5  # 효과 크기 (예: Cohen's d)
n = np.arange(10, 200, 5)  # 샘플 크기 범위
power = []

# 검정력 계산
for sample_size in n:
    # 귀무가설 분포
    null_mean = 0
    null_std = 1 / np.sqrt(sample_size)

    # 대립가설 분포
    alternative_mean = effect_size
    alternative_std = 1 / np.sqrt(sample_size)

    # 검정력 계산: 대립가설 하에서의 기각역
    critical_value = stats.norm.ppf(1 - alpha, loc=null_mean, scale=null_std)
    power_value = stats.norm.cdf(critical_value, loc=alternative_mean, scale=alternative_std)
    power.append(1 - power_value)

# 시각화
plt.figure(figsize=(10, 6))
plt.plot(n, power, label='검정력 곡선', color='blue')
plt.title('검정력 분석 결과')
plt.xlabel('샘플 크기 (n)')
plt.ylabel('검정력 (Power)')
plt.axhline(y=0.8, color='red', linestyle='--', label='검정력 0.8 기준선')
plt.axvline(x=n[np.argmax(np.array(power) >= 0.8)], color='green', linestyle='--',
            label='검정력 0.8에 도달하는 샘플 크기')
plt.legend()
plt.grid()
plt.show()
```

### ⭐tip.결과 해석 방법

1. 검정력 곡선

    - X축이 커짐에 따라 Y축의 검정력 값이 증가하는 것을 확인할 수 있다.
    
    - 즉, 샘플 크기가 커질수록 검정력이 증가한다.
    
    - 이는 샘플 크기가 클수록 대립가설이 참일 때 귀무가설을 기각할 가능성이 높아짐을 의미한다.

2. 검정력 0.8 기준선

    -  검정력 0.8 기준선은 검정력이 0.8인 지점을 시각적으로 나타낸다.
    
    - 이 기준선은 연구에서 일반적으로 요구하는 검정력 수준이다.

3. 검정력 0.8에 도달하는 샘플 크기

    - 검정력이 0.8에 도달하는 샘플 크기에서 수직선이 그려진다.
    
    - 이 지점은 검정력이 0.8 이상이 되기 위해 필요한 최소한의 샘플 크기를 나타낸다.
    
    - 그래프에서 이 지점을 보면, 검정력을 0.8로 설정할 때 필요한 샘플 크기를 알 수 있다.

### 7. 검정력 분석의 한계와 고려사항

1) 표본 크기 결정의 제약 사항

    - 표본 크기 결정은 비용, 시간 등의 제약을 받는다.
    
    - 즉, 자원이나 시간의 제한으로 인해 필요한 표본 크기를 확보하지 못할 수 있다.

    - 따라서 적절한 균형이 필요하다.

2) 효과 크기 과대/과소 추정 문제

    - 효과 크기를 부정확하게 추정하면 검정력 계산이 왜곡될 수 있다.
    
    - 따라서 신중한 추정이 필요하다.

3) 검정력 분석의 실제 적용 시 주의해야 할 사항

    - 이론적으로는 충분한 검정력을 확보했더라도, 데이터 수집 과정에서 발생하는 오류(데이터 품질)나 편향을 고려해야 한다.
    
    - 항상 검정력뿐만 아니라 전체 연구 설계를 검토하는 것이 중요하다.
